Aztertu beharreko sistema da $T$ tenperaturan dauden $m$ masako $N$ partikula bereiztezinez osatutako gas ideala. Sistemaren batezbesteko presioa da $p$. Partikula horietatik $N_{g}$ era askean higitzen ari dira $V=L_{g}^{3}$ bolumenean. Gainerakoak era askean higitzen ari dira $L_{s}^{2}$ azalerako gainazalean, xurgatuta. Esan bezala, $N=N_{g}+N_{s}$ betetzen da. Xurgatutako partikularen energia honako hau da: $\epsilon = \frac{|\vec{\bf{p}}^{2}|}{2m}-\epsilon_{0}$. Adierazpen horretan $\epsilon_{0}$ da partikulako xurgatze-energia. Lortu honako hauek:  

1. Partizio-funtzio klasikoa bi gasetarako: $Z_{g}$ eta $Z_{s}$.  
<br>  
1. Partikula *askez* (bolumenean) eta gainazalean xurgatutakoez  osatutako gasen Gibbs-en energia askeak: $G_{g}$ eta $G_{s}$.  
<br>  
1. Horien potentzial kimikoak: $\mu_{g}$ eta $\mu_{s}$.  
<br>  
1. Aipatutako $T$ tenperatura, bi *gas* horiek elkarren arteko  orekan daude. Lortu azalera unitateko xurgatutako partikulak eta  adeirazi baldintza esperimentalek definitzen dituzten parametroen  funtzioan: $T$, $p$...  
<br>  
1. $N$ eta $T$ konstante mantentzen dira baina gainazalean  xurgatuta ez dauden partikulek betetzen duten bolumena handitzen da.  
<br>  
   1. Zer eragina dauka horrek bi *gasen* arteko orekaren gainean? Xurgatuko al dira partikula gehiago?  
   
   1. Aztertu bolumenaren  aldaketaren eragina eta arrazoitu erantzuna.  
<br>     
1. Adierazi zein diren dimentsio bakarreko *gainazalean*, *orratzean* xurgatzen diren partikulez osatutako sistemari dagozkion $Z_{l}$, $G_{l}$ eta $\mu_{l}$.  
<br>  
1. Orratzean xurgatuko partikulak eta bolumean aske higituz doazenak elkarren arteko orekan daude. Baldintza horietan, zenbatekoa da dimentsio bakarrean eta bi dimentsioetan xurgatutako partikulen azalera *orokortu* unitatekoen arteko ratioa?
