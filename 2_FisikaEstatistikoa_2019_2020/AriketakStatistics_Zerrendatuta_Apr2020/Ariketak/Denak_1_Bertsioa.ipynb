`Kontuz`  

$N$ ohiko sare-puntu dago eta $N$ ez ohiko sare-puntu. Hots, $N$ atomo dugu eta $N$ akats izateko aukera:

Akatsen kopurua da $M$, $N$ baino askoz txikiagoa, baina oso handia ere bai. Datu horrek argitzen du Stirling-en hurbilketa erabili ahal izango dela bi zenbakien kasuan.

Zergatik ez da honako hau?

$$\Omega = \dfrac{N!}{M!\,(N-M)!}$$

eta, hortik, ondorioztatu entropia, honako hau,

$$S \approx 2\,k_{B}\, \left[ N\ln N - (N-M)\ln (N-M) - M\ln M \right]$$

1. Mikrokanonikoan:   
Kalkulatu behar da $(N,V_{1D},E=M\cdot \epsilon)$ hirukotearekin bateragarria den $\Omega$ mikroegoeren kopurua:  
    
    $$\Omega=\left[\frac{N!}{M!\cdot (N-M)!}\right]^{2}$$   
    - Adierazpen hori entropiaren $S=k_{B}\ln (\Omega)$ adeirazpenean *sartu* eta *Stirling-en* hurbilketa erabilita, erraztu $\Omega$   
    - ***Fisika sartu***:
        - gogorarazi $\frac{1}{T}\equiv \left(\frac{\partial S}{\partial E}\right)_{N,V}$

        - ondorioztatu $M=M(T)$

Kalkulatu behar da $Z_{M}(T,V_{1D})$ partizio-funtzioa:   
    
$$Z_{M}(T,V_{1D})=\left[\frac{N!}{M!\cdot (N-M)!}\right]^{2}\cdot e^{-\frac{E(=M\cdot\epsilon)}{k_{B}T}}$$

   <h2> Ebazpena   </h2>
   
Aintzakotzat hartu *multzo kanonikoan* ebatzi behar dela, beraz, partizio-funtzioa eraiki behar dugula.  

Apurka-apurka ebatziko da:  
1. Idatzi atomo bakarrari dagokion partizio-funtzioa. *Zenbat atal dauzka?*  
1. Idatzi horrelako $N_{a}$ atomoz osatutako multzoari dagokion energia-mailen sorta. *Zenbat atal dauzka?*  
1. Erreparatu honi buruz: modu bat baino gehiagotan kokatu daitezkeela xurgatutako atomoak gainazalean. Zenbat modutan? Jakina, endekapena dago.  
1. Lortu beharreko partizio-funtzioa, era formalean, honako hau da: $$Z(N_{s}, N_{a}, T) = Z_{Na}(N_{s}, T)$$
1. Azaldu adierazpena, konturatzeko aldagaien paperez.  
1. Honako hau da lortu beharko zenuken $F(N_{s}, N_{s}, T)$ Helmholtz-en energia askearen adierazpena:  
 
 $$F = - N_{a}\epsilon +k_{B}T\,\ln\left( 1- \dfrac{N_{a}}{N_{s}} \right) - k_{B}T\,N_{a}\ln \left(\dfrac{1-\dfrac{N_{a}}{N_{s}}}{\dfrac{N_{a}}{N_{s}}}\right)$$  
 
1. Bigarren galderari erantzuteko *Termon* egin genuena egin behar da. Gainazal batean xurgatutako atomoek osatzen duten multzoa da deskribatutako sistema. Demagun atomoak xurgatzeko iturria dela solidoaren lurruna, hots, demagun solidoa (gainazala) bere lurrunarekin ukipen *kimikoan* jartzen dela eta oreka lortzen dutela bi azpisistemek, solidoak eta lurrunak. Jakina, materia-trukearekiko lortuko dute oreka, onartuz tenperatura (eta presioa!) berean daudela. Horrek esan nahi du, bi *faseen* potentzial kimiko funtzioek balio berbera izango dutela, solidoarena eta lurrunarena hain zuzen ere. Orduan, itxitura-baldintza horrek finkatuko du baldintza esperimental horietan zenbatekoa izango den solidoak lurrunetik xurgatuko dituen atomoen ratioa.  
1. Jakina, lurruna gas idealtzat har dezakezu eta, beraz, problema hori ebatzita duzu.  
1. Idatzi gas idealari dagokion potentzial kimikoaren adierazpena eta berdindu aurreko ataletik abiatuta lortuko duzun potentzial kimikoaren adierazpenari.  
1. Lortuko duzun adierazpena honako hau da:   
 
 $$\dfrac{N_{a}}{N_{s}} = \left(k_{B}T\right)\,\cdot p\,\cdot \left[\lambda_{T}^{3}\right]\, e^{\dfrac{\epsilon}{k_{B}T}}$$  
 
1. Kontuan izan bolumenak ez duela parterik hartzen deskribatutako fisikan. Horren arabera, deskribatzen ari garen sistemaren askatasun-gradu mekanikoa, egotekotan, *izoztuta* dago. Horrexegatik onar dezakegu presioa konstantea dela ere bai.

Solido perfektu baten ereduan, $N$ atomo daude ohiko sare-puntuetan kokatuta. Ohiko sare-puntu horien arteko sare-puntuak, deskribatu nahi dugun solidoan, $N'$ dira. Bi kopuruak, $N$ eta $N'$ magnitude-ordena berekoak dira. Solido perfektuaren, ohiko sare-puntuak atomoz betetakoa bera, barne-energia da $U_{o}$.<linebreak>
Tenperatura finituan, $T$, atomo batek ohiko sare-puntu batetik ihes egin dezake, ez-ohiko sare-puntu batera, eta, beraz, hutsunea, (*vacancy*) utzi. Horrelako kasuan solidoa ez da perfektua eta *Frenkel-en* akatsa sortu dela esan ohi da. Akatsa sor daidin beharrezkoa den energia da $\epsilon>0$.  

Aztertuko duzun solidoa isolatuta dago eta $n$ *Frenkel-en* akats dauzka. Demagun $N\approx N'\ggg 1$ onargarria dela.  

Frogatu honako hau betetzen dela:  

$$n(T)\approx\sqrt{N\,N'}\,e^{-\dfrac{\epsilon}{2\,k_{B}T}} $$

   <h2> Ebazpena </h2>

**Multzo mikrokanonikoan**, dena konstante da, $E, N, V$, fisikoki sistema unibertsoa da, erabat isolatutako sistema.

1. Idatz itzazu itxitura baldintzak:   

   * partikula kopuru osoarena  
   * energia osoarena, har ezazu kontuan energietako bat negatiboa dela   
   
1. Erreparatu honi buruz: $N$ konstantea denez, maila bakoitzean dagoen partikula kopuruari dagokionez, aldagai bakarra dago: $n_{0}$ edo $n_{1}$.  

1. Egin honako aldagai-aldaketa. Definitu energia-mailetako partikula-frakzioa, i.e. $n_{i}^{'}\equiv\dfrac{n_{i}}{N}, i=1, 2$; eta energiaren itxitura-baldintzako ekuazioatik ondorioztatzen den $x$ aldagaia.  

1. Berridatzi energia-mailetako partikula-frakzioak definitu berri duzun $x$ aldagi adimentsionala erabiliz: $x\equiv\dfrac{E}{N\,\epsilon}$.  

1. Lortu $\Omega$ mikroegoeren kopurua, energia-mailetan dagaoen partikula kopuruak erabilita eta aplikatu *Stirling-en* hurbilketa. Honako hau lortu beharko zenuke:  

$$S = k_{B}\, \left[ N\ln N -n_{+}\ln n_{+} -n_{-}\ln n_{-} \right]$$    

----------  


**Multzo kanonikoaren kasuan**, abiapuntua termodinamika ondorioztatzeko da ***partizio-funtzioa***. Horixe da beraz, oraingo honetan lortu beharrekoa. Dena dela, ezer egin baino lehen, pentsatu apurtxo bat. Kontuara zaitez, sistemaren partikula osatzaileak identikoak izateaz gain, elkarrekintzarik gabekoak direla. Horrek esan nahi du, batari *gertatzen zaiona* beste edozeini gertatzen zaiola, denak aldi berean tangan sartuta egonik. Beraz, partikula baten kasua haintzakotzat hartuz, $N$ partikulez osatutako multzoari dagokiona lor dezakezu.  

1. Lortu partikula bakar bati dagokion partizio-funtzioa.  

   Jakina, definizioa ez da aldatzen, beraz, argi dago bi atal baino ez duela izango.  
   
1. Idatzi $N$ partikulez osatutako multzoari dagokion partizio-funtzioa, honako hau: $Q_{N}(V, T) = \sum_{r} \exp^{-\dfrac{E_{r}}{k_{B}\,T}}$  

1. Adierazpen horretan, $E_{r}$ da multzoari (ez multzo estatistikoari, kontuz) dagokion energia-maila posibleetako bat. Beraz, lortu berri duzun adierazpenetik honako hau idatzi beharko zenuke:   

    $$Q_{N}(V, T) = \left(Q_{1}(V, T)\right)^{N}$$  

1. Hau da, multzoaren partizio-funtzioa da $N$ aldiz partikula bakarrarena.  

1. Horren bidez, lortu multzo makrokanikoaren formalismoan oinarrizkoa den potentzial termodinamikoa, hots, $F(T, V, N)$, Helmholtz-en energia askea:   

    $$F(T, V, N) = - k_{B}T\,N\, \ln[Q_{1}(V, T)]$$  

1. Kontuan izan, entropiaz galdetu duela enuntziatua. Lortu entropia, $F$ potentzialetik abiatuta:   

    $$S = - \left(\dfrac{\partial F}{\partial T}\right)_{V, N}$$  

1. Aurreko puntuan lortutako adierazpena honako hau da, partikulako $S$ entropia, hain zuzen ere:   

    $$ \dfrac{S}{N} = k_{b} \left\{\ln \left[ e^{\dfrac{\epsilon}{k_{B}T}} + e^{\dfrac{\epsilon}{k_{B}T}} \right] + \dfrac{\epsilon}{k_{B}T}\, \left\{\dfrac{e^{\dfrac{\epsilon}{k_{B}T}} - e^{\dfrac{\epsilon}{k_{B}T}}}{e^{\dfrac{\epsilon}{k_{B}T}} + e^{\dfrac{\epsilon}{k_{B}T}}}\right\}\right\}$$  

---------  

**Parentesia**  

1. Berreskuratu mikrokanonikoaren formalismoan lortu duzun entropiaren adierazpena: $$ S = Nk_{B}\,\left[ \ln 2 - \frac{1}{2} (1+x)\ln (1+x) -\frac{1}{2}(1-x)\ln (1-x) \right] $$  

1. *Sartu* fisika multzo mikrokanonikoan, $\dfrac{1}{T}$ funtzioa kalkulatuz  

1. Kontuan izan honako hau:  

    $$ \dfrac{1}{T} = \left( \dfrac{\partial S}{\partial E} \right){V, N} = \left( \dfrac{\partial S}{\partial x} \right)\, \left(\dfrac{\partial x}{\partial E} \right){V,N}$$  
   
1. Frogatu hauxe betetzen dela:   

    $$ \dfrac{\epsilon}{k_{B}T} = \frac{1}{2}\,\ln\left(\dfrac{1-x}{1+x}\right)$$  

1. Erabili berdintza hori, eta lortu beharko duzun besteren bat, 6. ataleko partikulako entropiari dagokion adierazpen berria lortzeko.  

1. Pentsatu nola eraiki behar den multzoari dagokion energia partikulei dagozkien energiak erabilita.

   <h2> Ebazpena   </h2>

Gas ideala dagoeneko ebatzita dugu, beraz, enuntziatuko gas bakarra izango bagenu, dena eginda legoke. Baina hiru gas idealen arteko nahastura da aztertu behar dena. Onar dezagun hiru gas idea horiek osatzen duten nahastura berebat gas idela dela. Hortaz, gas ideala dugu, ebatzita dago...  

Gas idealaren problema ebazteko, betiko moduan, gas idealaren partikula osatzaieleei buruzko zenbait hipotesi onartu behar dugu. Izan ere, gas idealaren kasuan honako hauek: partikulak dira puntualak, identikoak eta elkarrekintzarik gabekoak (independenteak) eta bereiztezinak.  
 
1. *_Puntual_* izateak esan nahi du ez dutela barne-egiturarik, ez daukate barneko askatasun-gradurik eta, hortaz, askatasun-gradu translazionala baino ez dutela: hiru dimentsiotan (onartuko dugu) baino ez dute higitu.   

1. *_Identiko_* izateak esan du nahikoa dela partikula bakar bati dagokion problema ebaztea, besteen kasuan ebazpena berbera izango dugulako. Jakina, hori da elkarrekintzak daudela zein ez daudela, nahiz eta aztertu beharreko adibidean ez dagoen elkarrekintzarik.  

1. *_Elkarrekintzarik gabeko_* izateak esan nahi du, orain bai, bakarrari dagokiona ebatzita besteak ez daudela nahikoa dela problema ebazteko: ebazten ari garen problemako partikulak ez baititu *ikusiko* gainerakoak.  

1. Azkenik, *_bereiztezin_* izateak esan nahi du zenbatzea (edozein modutan eginda, mikroegoeren kopurua zuzenean, zein fase-espazioan) era berezian egin behar dela. Oraingoz, bereiztezintasuna kontuan hartzeko dugun tresna bakarra da *Gibbs*-ek berak erabilitakoa, hots, $N!$ faktorea. Baina kontuz, faktore hori erabiltzen denean baldintza esperimental jakinetan deskribatzen ari gara ebazten ari garen problema: baldintza *_klasikoetan_*. Horietan, batez bestean edozein egoeratan (energia-mailatan) dogoen partikula kopurua da 1 edo 0.  


Nahasturan ditugun hiru gas idealek aipatutako lau betekizunetatik lehenengo hirurak betetzen dituzte, nahasturari dagokionez; laurak, bakoitzari dagokionez. Hots, nahasturan dauden *kolore* ezberdineko partikulak *_bereizgarri_* dira eta, hortaz, nahasturan ez da bereiztasunaren faktorea kontuan hartu behar.  

Nahasturaren askatasun-graduak (translazionalak baino ez, jakina) $3 \cdot N = 3 \cdot (N_{gorri}+N_{urdin} +N_{berde}), N_{gorri}=N_{urdin}=N_{berde} = N$ dira, independenteak, askatasun-graduak baitira. Orduan, honako hau izango da nahasturari dagokion partizio-funtzioa:  

$$ Q_{N_{gorri}+N_{urdin} +N_{berde}}(V, T)_{nahastura} = Q_{N_{gorri}}(V, T)\, \times\,  Q_{N_{gorri}}(V,T)\, \times\, Q_{N_{berde}}(V, T)$$   

Hori guztia kontuan hartuz hauexekk dira lortu behar dituzun *_nahastura-entropiaren_* adierazpena lehenengo kasuan eta horrekiko aldea, bigarrenean:  

  - $$ S_{nahastura} = \dfrac{5}{3}(3N)\,k_{B}\,+ 3N\,k_{B}\, \ln\left(\dfrac{V\,T^{3/2}}{N\,\lambda_{T}^{3}}\right) $$   

  - $$ \Delta S = S_{nahastura} - S_{(3N)_{gorri}} = 3N\, k_{B}\, \ln(3) $$
