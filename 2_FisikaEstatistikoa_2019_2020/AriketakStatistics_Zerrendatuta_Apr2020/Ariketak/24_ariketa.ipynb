<div class="row alert alert-info"  role="alert">

   <br>

Hemoglobinako heme tokietan oxigeno molekulak kokatu beharrean, CO molekulak kokatzen badira, karbono monoxidoaren bidezko pozoitzea gertatzen da.   

Pozoitzea aztertzeko honako eredua erabil daiteke: heme tokitxoan $\textrm{O}_{2}$ edo CO molekulak egon daitezke, bata edo bestea. Lehenengoa dagoenean, hari dagokion energia $\epsilon_{A}$ da; aldiz, bigarrena dagoenean, $\epsilon_{B}$.   

1. Demagun $N$ heme tokitxo daudela $\textrm{O}_{2}$ eta CO molekulekin, gas fasean horiexek, orekan. Egoera horretan, $37^{\circ}$ Can, molekulen kontzentrazioek honako balio hauetan finkatu dituzte $\textrm{O}_{2}$ eta CO molekulen iheskortasunak: $z(\textrm{O}_{2})=1\times 10^{-5}$ eta $z(\textrm{CO})=1\times 10^{-7}$. Ez hartu kontuan spinaren endekapenak.  
<br>  
1. Lehenengo hurbilketa moduan, demagun ez dagoela CO molekularik.<breakline> Lortu $\epsilon_{A}$ energiaren balioa, heme taldeen %90 $\textrm{O}_{2}$ molekulez beteta egon daitezen. Adierazi emaitza eVetan, oxigeno molekulako.  
<br>  
1. Aurreko ataleko baldintzatan, oraingo honetan, CO molekulak ere badaude.<breakline> Lortu $\epsilon_{B}$ energiaren balioa heme talden %10 $\textrm{O}_{2}$ molekulez beteta egon daitezen.
