Aztertu beharreko sistema konposatua osatzen bi azpisistema dugu: $S_{1}$ eta $S_{2}$. Lehenak $N=3$ partikula ditu eta, bigarrenak $N=4$. Partikula horiek identikoak, independenteak eta bereizgarriak dira. Partikula horiek duten askatasun-gradua magnetikoa da; hots, $\mu$ momentu magnetikoaren bidez ezaugarrituta dago haien egoera magnetikoa. Gainera, egokia den ardatzean momentu magnetikoak daukan proiekzioak bi balio baino ez dauka: $\pm \mu$, kanpo eremu magnetikoarekin lerrokatuta, paraleloa eta antiparaleloa. Kanpo eremu magnetikoaren eta partikulen momentu magnetikoen arteko akoplamendu-energia ohikoa da: $E_{m}=-\bf{m}\cdot\bf{B}$.
