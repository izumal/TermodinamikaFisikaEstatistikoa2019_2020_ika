Gas-nahastura aztertu behar duzu, bi espeziez osatutakoa bera, A eta B.  

Nahastura gas ideala da eta nahasturakideen ($X$) presio partzialak honako hauek: $p_{X}$ ($X=A,B$). Gas-nahastura orekan dago $N$ puntu xurgatzaile daukan metalezko xaflarekin, $T$ tenperaturan. Xaflak $X$ motako partikula osatzailea (atomoa, molekula) xurgatzen duenean, haren energia $\epsilon_{X}$ da, gehienetan negatiboa, gas-egoeran aske higitzen ari diren gainontzeko partikulekiko.  

Xaflako puntu xurgatzaileek gehienez $A$ motako partikula bakarra har dezakete, bi egoeretan egon daitezkeenak; aldiz, $B$ motako bi partikula gehien jota; $B$ motako bakarra egoera bakarrean egon daiteke, baina bi daudenean, bi horiek 5 egoeretan egon daitezke. Gas-fasea oso handia da eta, beraz, partikulak xurgatzen direnean nekez nabaritzen da kontzentrazio-aldaketa: partikula-iturria da xaflarekiko, hortaz.<linebreak>
Lortu honako hauek:   

1. xaflaren azalean partikulen kontzentrazioa, hots, *estaldura*.  
<br>  
1. batez besteko $A$ eta $B$ partikulen kopuruak xaflan.
