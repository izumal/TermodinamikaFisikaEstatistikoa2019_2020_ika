Molekula elastiko luzea deskribatu daiteke $N$ loturez osatutako kate moduan. Lotura bakoitzaren egoera kuantikoak bi zenbaki kuantiko behar ditu: $n$ eta $l$. Loturen luzerak bi izan daitezke: $l=a$ eta $l=b$. Loturaren bibrazio-egoera osziladore harmoniko baten bidez adieraz daiteke, $\omega_{a}$ edo $\omega_{b}$ maiztasun angeluarrekoa bera, luzeraren arabera. Hots, honako hau da loturaren bibrazio-energia:   

$$
E(n,l)=(n+\dfrac{1}{2})\hbar\,\omega_{i},
\hspace{1cm} n=0,\dots,\infty,
\hspace{1cm}i=a,b
$$   

Molekula bera ($N$ loturezko katea) $T$ tenperaturan eta $\tau$ tentsiopean badago.    

Lortu honako hauek:    

1. batez besteko *energia* eta batez besteko katearen *luzera*    
<br>  
1. berebat, azter itzazu bi horien tenperatura handiko eta txikiko limiteak
