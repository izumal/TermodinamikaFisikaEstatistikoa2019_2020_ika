Aztertu beharreko sistema bi mailako $N$ atomoz, berezitezin eta elkarrekintzarik gabekoak haiek, osatuta dago. Honako bi energia-egoeretan egon daitezke atomoak: $E_{0}=0$ eta $E_{1}=\epsilon$. $E_{0}$ energia-mailako atomo kopurua $n_{0}$ da, eta $E_{1}$-ekoa, $n_{+1}$. Sistemaren barne-energia $U=n_{0}E_{0}+n_{1}E_{1}$ da. Lortu honako hauek:  

1. Sistemaren entropia, barne-energiaren funtzioan.  
<br>  
1. Sistemaren tenperatura. Zer baldintzatan izan daiteke negatibo?  
<br>

Kontura zaitez ariketa hau eta aurrekoa berdinak direla.   

Energiaren jatorria, normalean, definitu gabe dago, hobeto esateko, komeni zaigun balioan finkatu dezakegu. Orduan, berdin dio posibleak diren energia-mailak $(-\epsilon, \epsilon)$ izatea edo, $(E_{0}, E_{1})$ izatea. Jatorri-aldaketa egokiak *berdintzen* ditu energiak.  

*(Egin ezazu aldagai-aldaketa egokia eta moldatu aurreko ariketan lortutako emaitza honako ariketa honetara.)*
**Dena dela, ariketa honen helburua beste bat da.**   

1. Erreparatu bigarren galderari: sistemaren tenperatura lortu egin behar da!   
    Baina, zer ondorio dauka sistemaren tenperatura lortzeak? Hauxe da galdera ona. Sistemaren tenperatura lortzen bada, $S$ entropiaren adierazpena abiapuntuan izanik   
    
    $$\dfrac{1}{T}=\left(\dfrac{\partial S}{\partial E}\right)_{V, N}$$   
    
    tenperatura-funtzioa lortzen da.  
    
    Eman buelta funtzioari eta bakandu $E=E(T, V, N)$. Hori duzula, zer uste duzu lortuko duzula ordezkatzen baduzu energiaren adierazpen hori entropiaren adierazpenean, esaterako? Jakina, partizio-funtziotik abiatuta lortuko zenukeen adierazpena. Tenperaturaren adierazpenari buelta emandakoan lortzen denaz konturatzea oso inportantea da. Fisikoki sistemaren energia lortzen dugu tenperatuaren funtzioan...horixe da, hain zuzen ere, multzo kanonikoaren oinarria: sistemari aukera ematen zaio edozein energiarekin egoteko. Hori lortzen da bero-iturri batekin ukipen termikoan ezarrita. Bero-iturriak finkatzen dio teperatura sistemari. Eta, bide batez, *energia ere bai*: batezbesteko balioa, jakina. Proposatutako aldagai-aldaketa eginda, tenperatura sartzen dugu aldagai-independenteen sortan, beraz, multzo kanonikoaren formalismoan ari gara.   

1. Izan al daiteke sistema baten tenperatura negatibo? Ikusiko dugunez, zenbait baldintzatan (Pathria, "3.10 atala": "())", izan daiteke.
