Honako hau da aztertuko duzun problema: $\vec{\mu}$ momentu magnetikodun molekula diatomiko klasikoa. Momentu magnetikoa molekularen ardatzarekin lerrokatzen da $B$ kanpo eremu magnetikoaren eraginpean; eremua $z$ ardatzaren noranzko positiboan dago. Honako hau da molekularen hamiltondarra:   

$$
H=\dfrac{1}{2 I}\left(p_{\theta}^{2}+\dfrac{p_{\phi}^{2}}{\sin^{2}\theta}\right)-\mu B \cos\theta
$$   

Adierazpen horretan, $I$ da molekularen inertzia-momentua, $\mu=|\vec{\mu}|$, eta $-\mu B \cos\theta$ da molekularen momentu magnetikoaren eta kanpo eremu magnetikoaren arteko elkarrekintza-energia.    

Lortu honako hauek:    

1. Molekularen partizio-funtzioa.   
<br>  
1. Molekularen momentu magnetikoa.   
<br>  
1. $N$ molekulez osoatutako sistemaren energia.
