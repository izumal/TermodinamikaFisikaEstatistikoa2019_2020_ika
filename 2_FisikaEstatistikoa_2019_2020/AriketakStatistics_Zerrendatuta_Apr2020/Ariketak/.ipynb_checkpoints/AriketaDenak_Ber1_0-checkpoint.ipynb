Sare batek $N$ ohiko sare-puntu ditu eta $N$ sare-puntuen arteko puntuak. Sare-puntuak denak berezigarri dira (jakina, sare-puntuen artekoak ere bai). $N$ atomo identiko jartzen dira sarean: $M$, sare-puntuen arteko puntuetan eta $N-M$, ohiko sare-puntuetan ($N\gg M\gg 1$). Ohiko sare-puntuan dagoen atomoaren energia $E=0$ da eta sare-puntuen arteko puntuan dagoen atomoarena, $E=\epsilon$.   

Lor itzazu, (multzo estatistiko) ***mikrokanonikoan*** eta *kanonikoan*, sare horri dagozkion barne-energia ($E$) eta bero-ahalmena ($C_{1D-V}$), $T$ tenperaturaren funtzioan.

Aztertu ondo deskribatutako eredua eta konturatu ez dela bi mailako sistema. Enuntziatuak eskatzen dituen magnitudeak ondorioztatzeko, multzo mikrokanonikoa erabiliko dugu, hots, mikroegoeren kopurua kalkulatu behar dugu.  

`Kontuz`  

$N$ ohiko sare-puntu dago eta $N$ ez ohiko sare-puntu. Hots, $N$ atomo dugu eta $N$ akats izateko aukera:

Akatsen kopurua da $M$, $N$ baino askoz txikiagoa, baina oso handia ere bai. Datu horrek argitzen du Stirling-en hurbilketa erabili ahal izango dela bi zenbakien kasuan.  

1. Idatzi mikroegoeren kopurua:  
Zergatik ez da honako hau?

$$\Omega = \dfrac{N!}{M!\,(N-M)!}$$

2. Lortu mikroegoeren kopurua:   
eta, hortik, ondorioztatu entropia, honako hau,

$$S \approx 2\,k_{B}\, \left[ N\ln N - (N-M)\ln (N-M) - M\ln M \right]$$

1. Mikrokanonikoan:   
Kalkulatu behar da $(N,V_{1D},E=M\cdot \epsilon)$ hirukotearekin bateragarria den $\Omega$ mikroegoeren kopurua:  
    
    $$\Omega=\left[\frac{N!}{M!\cdot (N-M)!}\right]^{2}$$   
    - Adierazpen hori entropiaren $S=k_{B}\ln (\Omega)$ adeirazpenean *sartu* eta *Stirling-en* hurbilketa erabilita, erraztu $\Omega$   
    - ***Fisika sartu***:
        - gogorarazi $\frac{1}{T}\equiv \left(\frac{\partial S}{\partial E}\right)_{N,V}$
        - kalkulatu $\frac{1}{T}$ aurreko adierazpenenetik
        - ondorioztatu $M=M(T)$

2. Kanonikoan:   
Kalkulatu behar da $Z_{M}(T,V_{1D})$ partizio-funtzioa:   
    
$$Z_{M}(T,V_{1D})=\left[\frac{N!}{M!\cdot (N-M)!}\right]^{2}\cdot e^{-\frac{E(=M\cdot\epsilon)}{k_{B}T}}$$

Berdindu azalean xurgatutako atomoen eta gasaren atomoen potentzial kimikoak, oreka kimikoan baitaude.  

Kasu honetan, solido perfektuaren azala bi dimentsioko sistematzat hartuko da. Sare-puntuetako puntuek atomoak xurga ditzakete eta xurgatutako atomoek osatuko dute aztertu nahi dugun sistema: solidoa bera ez da gure intereseko sistema, bere azalean xurgatu dituen atomoek osatzen dutena baino. Gainera, enuntziatuaren arabera, xurgatutako atomoak *erosoago* daude xurgaturik, haiekin lotutako energia $-\epsilon$ baita.  

Gainazal xurgatzailean dagoen sare-puntuen kopurua da $N_{s}$ eta xurgatutako atomoen kopurua, $N_{a}$, aurreko kopurua baino askoz txikiagoa, baina biak dira oso-oso handiak, hots, bien kasuan erabil daiteke *Stirling*-en hurbilketa.  

Xurgatutako atomoak identikoak eta independenteak dira. Hau da, bakar bati dagokion problema ebatzita bera bakarrik egonda, gainazalean bera bakarraik xurgatuta dagoela edo beste batekin edo beste denekin, emaitza berbera lortzen da. Jakina, emaitza dagoeneko ezaguna da: enuntziatuaren arabera, xurgatutako atomoaren energia $-\epsilon$ da. Horixe da sistemaren partikula osatzaileen problema ebatzitakoan hari dagozkion energia-mailen sorta! Kasu honetan, energia-maila bakarra.

   <h2> Ebazpena   </h2>
   
Aintzakotzat hartu *multzo kanonikoan* ebatzi behar dela, beraz, partizio-funtzioa eraiki behar dugula.  

Apurka-apurka ebatziko da:  
1. Idatzi atomo bakarrari dagokion partizio-funtzioa. *Zenbat atal dauzka?*  
1. Idatzi horrelako $N_{a}$ atomoz osatutako multzoari dagokion energia-mailen sorta. *Zenbat atal dauzka?*  
1. Erreparatu honi buruz: modu bat baino gehiagotan kokatu daitezkeela xurgatutako atomoak gainazalean. Zenbat modutan? Jakina, endekapena dago.  
1. Lortu beharreko partizio-funtzioa, era formalean, honako hau da: $$Z(N_{s}, N_{a}, T) = Z_{Na}(N_{s}, T)$$
1. Azaldu adierazpena, konturatzeko aldagaien paperez.  
1. Honako hau da lortu beharko zenuken $F(N_{s}, N_{s}, T)$ Helmholtz-en energia askearen adierazpena:  
 
 $$F = - N_{a}\epsilon +k_{B}T\,\ln\left( 1- \dfrac{N_{a}}{N_{s}} \right) - k_{B}T\,N_{a}\ln \left(\dfrac{1-\dfrac{N_{a}}{N_{s}}}{\dfrac{N_{a}}{N_{s}}}\right)$$  
 
1. Bigarren galderari erantzuteko *Termon* egin genuena egin behar da. Gainazal batean xurgatutako atomoek osatzen duten multzoa da deskribatutako sistema. Demagun atomoak xurgatzeko iturria dela solidoaren lurruna, hots, demagun solidoa (gainazala) bere lurrunarekin ukipen *kimikoan* jartzen dela eta oreka lortzen dutela bi azpisistemek, solidoak eta lurrunak. Jakina, materia-trukearekiko lortuko dute oreka, onartuz tenperatura (eta presioa!) berean daudela. Horrek esan nahi du, bi *faseen* potentzial kimiko funtzioek balio berbera izango dutela, solidoarena eta lurrunarena hain zuzen ere. Orduan, itxitura-baldintza horrek finkatuko du baldintza esperimental horietan zenbatekoa izango den solidoak lurrunetik xurgatuko dituen atomoen ratioa.  
1. Jakina, lurruna gas idealtzat har dezakezu eta, beraz, problema hori ebatzita duzu.  
1. Idatzi gas idealari dagokion potentzial kimikoaren adierazpena eta berdindu aurreko ataletik abiatuta lortuko duzun potentzial kimikoaren adierazpenari.  
1. Lortuko duzun adierazpena honako hau da:   
 
 $$\dfrac{N_{a}}{N_{s}} = \left(k_{B}T\right)\,\cdot p\,\cdot \left[\lambda_{T}^{3}\right]\, e^{\dfrac{\epsilon}{k_{B}T}}$$  
 
1. Kontuan izan bolumenak ez duela parterik hartzen deskribatutako fisikan. Horren arabera, deskribatzen ari garen sistemaren askatasun-gradu mekanikoa, egotekotan, *izoztuta* dago. Horrexegatik onar dezakegu presioa konstantea dela ere bai.

   <h2> Solidoak: bi dimentsioko solidoa (2). <em>Frenkel-en</em> akatsak   </h2> 

   <h2> Helburua:    </h2>
   
   
   <hr>

   <h2> Ikasiko duzuna:  </h2>  
   
   
</div>

<div class="row alert alert-info"  role="alert">
   <br>
   <br>

Solido perfektu baten ereduan, $N$ atomo daude ohiko sare-puntuetan kokatuta. Ohiko sare-puntu horien arteko sare-puntuak, deskribatu nahi dugun solidoan, $N'$ dira. Bi kopuruak, $N$ eta $N'$ magnitude-ordena berekoak dira. Solido perfektuaren, ohiko sare-puntuak atomoz betetakoa bera, barne-energia da $U_{o}$.<linebreak>
Tenperatura finituan, $T$, atomo batek ohiko sare-puntu batetik ihes egin dezake, ez-ohiko sare-puntu batera, eta, beraz, hutsunea, (*vacancy*) utzi. Horrelako kasuan solidoa ez da perfektua eta *Frenkel-en* akatsa sortu dela esan ohi da. Akatsa sor daidin beharrezkoa den energia da $\epsilon>0$.  

Aztertuko duzun solidoa isolatuta dago eta $n$ *Frenkel-en* akats dauzka. Demagun $N\approx N'\ggg 1$ onargarria dela.  

Frogatu honako hau betetzen dela:  

$$n(T)\approx\sqrt{N\,N'}\,e^{-\dfrac{\epsilon}{2\,k_{B}T}} $$

   <h2> Solidoak: bi dimentsioko solidoa (2). <em>Schottky-ren</em> akatsak   </h2> 

   <h2> Helburua:    </h2>
   
   
   <hr>

   <h2> Ikasiko duzuna:  </h2>  
   
   
</div>

<div class="row alert alert-info"  role="alert">
   <br>
   <br>

Aztertu beharreko sistema da $N$ atomo identikoek eratzen duten solidoa, hiru dimentsioko banaketa erregularra. Atomo horietako $n$ ($1\gg n\gg N$) dauden sare-puntuetatik kendu eta solidoaren gainazaleko sare-puntuetan ezartzen badira, *Schottky-ren* motako akatsadun solido eratuko da, $n$ akatsdun solidoa.    

Demagun $\epsilon$ dela beharrezkoa den energia *Schottky-ren* motako akatsa sorrarazteko, hots, beharrezkoa den energia (solidoaren barneko) sare-puntu batean dagoen atomoa gainazaleko sare-puntu batera pasarazteko.  

>Frogatu honako adierazpen hurbildua egokia dela oreka-egoeran, $\epsilon \gg k_{B}T$ betetzen duen $T$ tenperaturan,  
><br>
>$$\boxed{n \approx N \exp^{-\frac{\epsilon}{k_{B}T}} }$$  
<br>
> Onartu akatsak sorrarazten direnean, gertatzen den bolumen-aldaketa baztergarria dela.

   <h2> Ebazpena </h2>

**Multzo mikrokanonikoan**, dena konstante da, $E, N, V$, fisikoki sistema unibertsoa da, erabat isolatutako sistema.

1. Idatz itzazu itxitura baldintzak:   

   * partikula kopuru osoarena  
   * energia osoarena, har ezazu kontuan energietako bat negatiboa dela   
   
1. Erreparatu honi buruz: $N$ konstantea denez, maila bakoitzean dagoen partikula kopuruari dagokionez, aldagai bakarra dago: $n_{0}$ edo $n_{1}$.  

1. Egin honako aldagai-aldaketa. Definitu energia-mailetako partikula-frakzioa, i.e. $n_{i}^{'}\equiv\dfrac{n_{i}}{N}, i=1, 2$; eta energiaren itxitura-baldintzako ekuazioatik ondorioztatzen den $x$ aldagaia.  

1. Berridatzi energia-mailetako partikula-frakzioak definitu berri duzun $x$ aldagi adimentsionala erabiliz: $x\equiv\dfrac{E}{N\,\epsilon}$.  

1. Lortu $\Omega$ mikroegoeren kopurua, energia-mailetan dagaoen partikula kopuruak erabilita eta aplikatu *Stirling-en* hurbilketa. Honako hau lortu beharko zenuke:  

$$S = k_{B}\, \left[ N\ln N -n_{+}\ln n_{+} -n_{-}\ln n_{-} \right]$$    

----------  


**Multzo kanonikoaren kasuan**, abiapuntua termodinamika ondorioztatzeko da ***partizio-funtzioa***. Horixe da beraz, oraingo honetan lortu beharrekoa. Dena dela, ezer egin baino lehen, pentsatu apurtxo bat. Kontuara zaitez, sistemaren partikula osatzaileak identikoak izateaz gain, elkarrekintzarik gabekoak direla. Horrek esan nahi du, batari *gertatzen zaiona* beste edozeini gertatzen zaiola, denak aldi berean tangan sartuta egonik. Beraz, partikula baten kasua haintzakotzat hartuz, $N$ partikulez osatutako multzoari dagokiona lor dezakezu.  

1. Lortu partikula bakar bati dagokion partizio-funtzioa.  

   Jakina, definizioa ez da aldatzen, beraz, argi dago bi atal baino ez duela izango.  
   
1. Idatzi $N$ partikulez osatutako multzoari dagokion partizio-funtzioa, honako hau: $Q_{N}(V, T) = \sum_{r} \exp^{-\dfrac{E_{r}}{k_{B}\,T}}$  

1. Adierazpen horretan, $E_{r}$ da multzoari (ez multzo estatistikoari, kontuz) dagokion energia-maila posibleetako bat. Beraz, lortu berri duzun adierazpenetik honako hau idatzi beharko zenuke:   

    $$Q_{N}(V, T) = \left(Q_{1}(V, T)\right)^{N}$$  

1. Hau da, multzoaren partizio-funtzioa da $N$ aldiz partikula bakarrarena.  

1. Horren bidez, lortu multzo makrokanikoaren formalismoan oinarrizkoa den potentzial termodinamikoa, hots, $F(T, V, N)$, Helmholtz-en energia askea:   

    $$F(T, V, N) = - k_{B}T\,N\, \ln[Q_{1}(V, T)]$$  

1. Kontuan izan, entropiaz galdetu duela enuntziatua. Lortu entropia, $F$ potentzialetik abiatuta:   

    $$S = - \left(\dfrac{\partial F}{\partial T}\right)_{V, N}$$  

1. Aurreko puntuan lortutako adierazpena honako hau da, partikulako $S$ entropia, hain zuzen ere:   

    $$ \dfrac{S}{N} = k_{b} \left\{\ln \left[ e^{\dfrac{\epsilon}{k_{B}T}} + e^{\dfrac{\epsilon}{k_{B}T}} \right] + \dfrac{\epsilon}{k_{B}T}\, \left\{\dfrac{e^{\dfrac{\epsilon}{k_{B}T}} - e^{\dfrac{\epsilon}{k_{B}T}}}{e^{\dfrac{\epsilon}{k_{B}T}} + e^{\dfrac{\epsilon}{k_{B}T}}}\right\}\right\}$$  

---------  

**Parentesia**  

1. Berreskuratu mikrokanonikoaren formalismoan lortu duzun entropiaren adierazpena: $$ S = Nk_{B}\,\left[ \ln 2 - \frac{1}{2} (1+x)\ln (1+x) -\frac{1}{2}(1-x)\ln (1-x) \right] $$  

1. *Sartu* fisika multzo mikrokanonikoan, $\dfrac{1}{T}$ funtzioa kalkulatuz  

1. Kontuan izan honako hau:  

    $$ \dfrac{1}{T} = \left( \dfrac{\partial S}{\partial E} \right){V, N} = \left( \dfrac{\partial S}{\partial x} \right)\, \left(\dfrac{\partial x}{\partial E} \right){V,N}$$  
   
1. Frogatu hauxe betetzen dela:   

    $$ \dfrac{\epsilon}{k_{B}T} = \frac{1}{2}\,\ln\left(\dfrac{1-x}{1+x}\right)$$  

1. Erabili berdintza hori, eta lortu beharko duzun besteren bat, 6. ataleko partikulako entropiari dagokion adierazpen berria lortzeko.  

1. Pentsatu nola eraiki behar den multzoari dagokion energia partikulei dagozkien energiak erabilita.

   <h2> Ebazpena   </h2>

Gas ideala dagoeneko ebatzita dugu, beraz, enuntziatuko gas bakarra izango bagenu, dena eginda legoke. Baina hiru gas idealen arteko nahastura da aztertu behar dena. Onar dezagun hiru gas idea horiek osatzen duten nahastura berebat gas idela dela. Hortaz, gas ideala dugu, ebatzita dago...  

Gas idealaren problema ebazteko, betiko moduan, gas idealaren partikula osatzaieleei buruzko zenbait hipotesi onartu behar dugu. Izan ere, gas idealaren kasuan honako hauek: partikulak dira puntualak, identikoak eta elkarrekintzarik gabekoak (independenteak) eta bereiztezinak.  
 
1. *_Puntual_* izateak esan nahi du ez dutela barne-egiturarik, ez daukate barneko askatasun-gradurik eta, hortaz, askatasun-gradu translazionala baino ez dutela: hiru dimentsiotan (onartuko dugu) baino ez dute higitu.   

1. *_Identiko_* izateak esan du nahikoa dela partikula bakar bati dagokion problema ebaztea, besteen kasuan ebazpena berbera izango dugulako. Jakina, hori da elkarrekintzak daudela zein ez daudela, nahiz eta aztertu beharreko adibidean ez dagoen elkarrekintzarik.  

1. *_Elkarrekintzarik gabeko_* izateak esan nahi du, orain bai, bakarrari dagokiona ebatzita besteak ez daudela nahikoa dela problema ebazteko: ebazten ari garen problemako partikulak ez baititu *ikusiko* gainerakoak.  

1. Azkenik, *_bereiztezin_* izateak esan nahi du zenbatzea (edozein modutan eginda, mikroegoeren kopurua zuzenean, zein fase-espazioan) era berezian egin behar dela. Oraingoz, bereiztezintasuna kontuan hartzeko dugun tresna bakarra da *Gibbs*-ek berak erabilitakoa, hots, $N!$ faktorea. Baina kontuz, faktore hori erabiltzen denean baldintza esperimental jakinetan deskribatzen ari gara ebazten ari garen problema: baldintza *_klasikoetan_*. Horietan, batez bestean edozein egoeratan (energia-mailatan) dogoen partikula kopurua da 1 edo 0.  


Nahasturan ditugun hiru gas idealek aipatutako lau betekizunetatik lehenengo hirurak betetzen dituzte, nahasturari dagokionez; laurak, bakoitzari dagokionez. Hots, nahasturan dauden *kolore* ezberdineko partikulak *_bereizgarri_* dira eta, hortaz, nahasturan ez da bereiztasunaren faktorea kontuan hartu behar.  

Nahasturaren askatasun-graduak (translazionalak baino ez, jakina) $3 \cdot N = 3 \cdot (N_{gorri}+N_{urdin} +N_{berde}), N_{gorri}=N_{urdin}=N_{berde} = N$ dira, independenteak, askatasun-graduak baitira. Orduan, honako hau izango da nahasturari dagokion partizio-funtzioa:  

$$ Q_{N_{gorri}+N_{urdin} +N_{berde}}(V, T)_{nahastura} = Q_{N_{gorri}}(V, T)\, \times\,  Q_{N_{gorri}}(V,T)\, \times\, Q_{N_{berde}}(V, T)$$   

Hori guztia kontuan hartuz hauexekk dira lortu behar dituzun *_nahastura-entropiaren_* adierazpena lehenengo kasuan eta horrekiko aldea, bigarrenean:  

  - $$ S_{nahastura} = \dfrac{5}{3}(3N)\,k_{B}\,+ 3N\,k_{B}\, \ln\left(\dfrac{V\,T^{3/2}}{N\,\lambda_{T}^{3}}\right) $$   

  - $$ \Delta S = S_{nahastura} - S_{(3N)_{gorri}} = 3N\, k_{B}\, \ln(3) $$

# &#9989; 26 Ariketa: Multzo kanonikoa eta Multzo Makrokanonikoa   

## Sistema txikiak 7
## Solidoak eta gasak (1): *Gainazal xurgatzailean molekulak polarizatzen dira*

   
## Helburua:  
   
   1. Konturatzea multzo estatistikoen formalismoek baldintzatzen dutela sistematzat aztertu beharrekoa  
   
   2. Konturatzea nahiz eta bi multzo estatistiko ezberdin erabili, limite asintotikoan, biek ematen dutela emaitza berbera  
   
   3. *Limite asintotikoa* praktikoki da Stirling-en hurbilketa erabiltzea  
   
   4. Barneko askatasun-graduak edozein testuingurutan *erabil/ager* daitezke, kasu honetan, elketrikoa
   
##  Ikasiko duzuna:    

   1. Puntu xurgatzialea aipatzen denean, normalean, erabili beharreko multzo estatistikoa makrokanonikoa da.  
      Puntu xurgatzaile horiek osatuko dute *sistema osoa*.  
      Puntu xurgatzialeak *identiko* eta *independente* izanda, bakarra aztertzearekin nahikoa da.  
      Puntu xurgatzaileek edo ez daukate partikularik edo partikula(k) xurgatzen dituzte, enuntziatuaren arabera.  
   <br>  
   2. Puntu xurgatzaileez osatutako sistema osoa aztertu daiteke multzo estatistiko kanonikoaren formalismoa erabilita.  
      Horrelako sistema deskribatzeko *partikulak xurgatuta daudela dagozkien energia-maialak* hartu behar dira kontuan; hots, ***eredu kanonikoan*** partikulak puntu xurgatzaieleetan daude.  
   <br>       
   3. *Toki/puntu* xurgatzaileak **beti** dira **bereizgarri** (nahiz eta partikula bereiztezinez beteta egon...)


   
   
</div>

<div class="row alert alert-info"  role="alert">
   <br>
   <br>

Gainazal batean, $M$ tokitxo xurgatzaile ditu berak, $m$ masako $n$ partikula lotuta daude ($n$ kopurua oso-oso handia da). Tokitxo xurgatzaileek atomo bakarra har dezakete eta, hartutakoan, $AB$ molekula eratzen da. Beraz, pentsa daiteke, tokitxo xurgatzaileetan $B$ motako partikulak daudela. Eratutako molekulek badute momentu dipolarra, $\vec{p}$, bertikalki modu bakarrean orientatu daitekeena eta, horizontalki, lau modutan: $p_{z}$ eta $\pm\,p_{x}$ eta $\pm\,p_{y}$, hurrenez hurren. Lotura-energia da $\epsilon_{0}$. Demagun kanpo eremu elektrikoa ezarri dela, bertikalki: $\vec{\mathscr{\epsilon}}$. Jakina, ezarritako eremua elektrikoaren eta momentu dipolarraren arteko akoplamendu-energia da $E = - \vec{p}\cdot\vec{\Large{\varepsilon}}$. Azkenik, sistema osoaren polarizazio honako lanaren adierazpenaren bidez definituta dago: $dW = - \vec{P}\cdot d\vec{\Large{\varepsilon}}$.  
 
1. Lortu sistemaren partizio-funtzioa: $Z_{n}(T)$  
<br>  
1. Lortu potentzial kimikoaren adierazpena: $\mu(T;n)$  
<br>  
1. Ondorioztatu zenbatekoa den estaltzea: $\langle n \rangle$.  
  Kasu honetan, aurreko adierazpenetik abiatu beharko duzu estaltzea, potentzial kimikoaren funtzioan idazteko, hau da, onartuz, ezaguna dela potentzial kimikoa  
<br>  
1. Ber-lortu $\langle n \rangle$ estaltzearen adierazpena partizio-funtzio gran kanonikoa erabiliz  
<br>

Xaflak $X$ motako partikula osatzailea (atomoa, molekula) xurgatzen duenean, haren energia $\epsilon_{X}$ da, gehienetan negatiboa, gas-egoeran aske higitzen ari diren gainontzeko partikulekiko.

1. xaflaren azalean partikulen kontzentrazioa, hots, *estaldura*.

## Sistema txikiak 9
## Solidoak eta gasak (3): *Gainazal xurgatzailea eta iturriak*

   
## Helburua:  
   
   1. Konturatzea nola aldatzen diren gauzak puntu xurgatzaileen ezaugarrien arabera: partikulen oinarrizko izaeraren ***joko-eredua*** (*toy model*).     

   2. Egitea kalkulu kuantitatiboak, *eskuz bada ere*, sistema txikietan.     

   3. Konturatzea, kuantitatiboki eta kualitatiboki, nola eraikitzen diren sistema *handietako* energia-mailak orbitalen energiak erabilita.
   
   4. Konturatzea, aurrekoa eta toki xurgatzaileen betetze-ezaugarriak kontuan hartuta, nola aldatzen diren sistema osoari dagozkion energia-maialen sorta.
   
      
##  Ikasiko duzuna:    

   1. Enuntziatuaren arabera, ez dago argi partikulak bereiztezin ala bereigarri diren, konturatu.  
      Ebatzi modu batean eta bestean, proposamena.    
   <br>  
   2. Batzuetan, partikulen *benetako* izaera zein den jakin gabe, betetzeko baldintzek finkatzen diete izaera, gutxienez, akademikoa. Aldiz, edozein kopurutan sar badaitezke partikulak, ***bosoi-izaerakotzat*** jo daitezke.       
   <br>      
   3. Sistema txikietan eta partikula kopuruak txikiak direnean, kalkuluak erraz egin daitezke, *eskuz zenbatuz*.   
   <br>   
   4. Puntu xurgatzaileen inguruko ariketak *akademikoki* ebazteko ez da beharrezkoa aztergai den sistemarekin oreka kimikoan dagoen materia-iturria.

- Gas idealaren Multzo Makrokanonikoaren Partizio-funtzioa:

   $$\mathscr{Z}(T,V,\mu) = \sum\limits^{\infty}_{N=0}  z^{N} \left[\frac{1}{N!}\right] \left[Z_{1}\right]^{N} \Rightarrow \mathscr{Z}(T,V,\mu) = \exp[z\cdot Z_{1}]  $$

   - adierazpen horretako $N$ partikula kopuruaren balio bakoitzeko partizio-funtzio kanonikoa, $Z_{N}$ modu honetan eraiki da: $Z_{N}\equiv \left[Z_{1}\right]^{N}$, partikula osatzaileak identikoak eta independenteak baitira
   - gainera, berezitezinak direnez, **baina oraindik ez *kuantikoak***, Gibbs-en *ad-hoc* faktorea erabili da bereizezintasuna kontuan hartzeko: $\frac{1}{N!}$
   - kasu jakin honetan, oso *erraza* da partizio-funtzio makrokanonikoa:

   $$\mathscr{Z}_{N}(T,V,\mu)= \exp[z\cdot Z_{1}(T,V)]$$

   - eta potentzial termodinamiko egokia, beraz, honako hau da $pV=\ln\mathscr{Z}_{N}$:

     $$\boxed{pV=z\cdot Z_{1}(T,V)}$$

   - Lehenengo dierazpenetik edozer lor daiteke, besteak beste, $\langle N\rangle$ batez besteko partikula kopurua:

   $$\langle N\rangle \equiv N(z, V, T)=z\left[\frac{\partial}{\partial z} \ln \mathscr{Z}(z, V, T)\right]_{V, T}=k T\left[\frac{\partial}{\partial \mu} \ln \mathscr{Z}(\mu, V, T)\right]_{V, T}\Rightarrow \langle N\rangle = z\cdot Z_{1}$$

   - definizioak erabilita:
       - $z=\exp\left(\frac{\mu}{k_{B}T}\right)$
       - $Z_{1}(T,V)=\frac{V}{\lambda^{3}_{T}}\cdot \left[Z^{\textrm{barne}}_{1}\right]$, geroxeago erabiliko da
   - potentzial kimikoaren adeirazpena ondorioztatu daiteke:

   $$\langle N\rangle = \exp\left(\frac{\mu}{k_{B}T}\right) Z_{1}(T,V)\Rightarrow \boxed{\mu = k_{B}T \ln\left(\frac{\langle N \rangle}{Z_{1}}\right)}$$

- Erreakzionatzen dauden bi *gasak*, erreaktiboak eta produktuak, gas egoeran daude eta gas idealtzat har daitezke, beraz, goiko guztia baliagarrai da erreaktiboen eta produktuen kasuetan
- Gainera, *orekan daude* enuntziatuaren arabera, jakina, erreakzio kimikoaren $K(T)$ oreka-konstantea lortu behar da, *oreka*...hortaz, onartuko da erreaktiboak eta produktuak oreka kimikoan daudela eta, ondorioz, haien potentzial kimiko *pisatuak*, erreakzioaren ekuazioko, $a\,A \rightleftharpoons b\,B$ , koefiziente estekiometrikoak dira haien pisuak eta, azkenik, honako hau da erreakzio kimikoaren oreka kimikoa adierazten duen berdintza:

    $$a\mu_{A} = b\mu_{B}$$

    - potentizal kimikoaren adierazpena berdintzan ordezkatuta honako hau lortzen da:

    $$a\cdot k_{B}T \ln\left(\frac{\langle N_{A} \rangle}{Z^{A}_{1}}\right) = b\cdot k_{B}T \ln\left(\frac{\langle N_{B} \rangle}{Z^{B}_{1}}\right)$$

    $$\ln\left[\left( \frac{\langle N_{A} \rangle}{Z^{A}_{1}}\right)\right]^{a} = \ln\left[\left( \frac{\langle N_{B} \rangle}{Z^{B}_{1}}\right)\right]^{b}$$

    $$\left( \frac{\langle N_{A} \rangle}{Z^{A}_{1}}\right)^{a} = \left( \frac{\langle N_{B} \rangle}{Z^{B}_{1}}\right)^{b}\Rightarrow \frac{\langle N_{A} \rangle^{a}}{\left(Z^{A}_{1}\right)^{a}}= \frac{\langle N_{B} \rangle^{b}}{\left(Z^{B}_{1}\right)^{b}}$$

    - $K(T)$ erreakzio konstantearen definizioa honako hau da: $K(T)\equiv \frac{n^{b}_{B}}{n^{a}_{A}}$, hots espezioekin lotutako $n\equiv \frac{\langle N \rangle}{V}$ partikula-dentsitateen arteko zatidura *pisatua*, pisuak dira erreakzio-ekuazioko koefiziente estekiometrikoak, berretura moduan agertuko direnak erreakzio-konstantean
    - partikula-dentsitatearen definiziotik lor daiteke batez besteko partikula kopurua: $\langle N\rangle = nV$
    - eta adierazepn hori ordezkatu daiteke *orekaren berdinatzaren bukaerako adierazpenean*:

    $$\frac{\left( n_{A}V \right)^{a}}{\left(Z^{A}_{1}\right)^{a}}= \frac{\left( n_{B}V \right)^{b}}{\left(Z^{B}_{1}\right)^{b}}\Rightarrow \frac{n^{b}_{B}}{n^{a}_{A}}= \frac{\left(Z^{B}_{1}\right)^{b}}{\left(Z^{A}_{1}\right)^{a}}\frac{V^{a}}{V^{b}}\Rightarrow \boxed{K(T) = \frac{\left(Z^{B}_{1}\right)^{b}}{\left(Z^{A}_{1}\right)^{a}}V^{a-b}}$$

- Aztertu beharreko aplikazioaren kasuan honako hau da erreakzioa:

    $$\textrm{O}_{2}\,\rightleftharpoons\,2\,\textrm{O}$$

    - $a=1$, $b=2$
    - $A=\textrm{O}_{2}$, $B=\textrm{O}$
    - eta $K(T)$ konstantea da:

    $$\boxed{K_{\textrm{O}_{2}\,\rightleftharpoons\,2\,\textrm{O}}(T)=\frac{\left(Z^{\textrm{O}}_{1}\right)^{2}}{\left(Z^{\textrm{O}_{2}}_{1}\right)^{1}}V^{-1}}$$

    - bestetik, partikula osatzaileen partizio-funtzioen adierazpenak honako hauek dira:
        - O, monoatomikoa denez, ez dauka barneko askatasun-graduarik, spina salbo
          askatasun-gradu bakarra da *translazionala* eta partizio-funtzioa honako hau:
          $Z^{\textrm{O}}_{1}= 4\cdot \frac{V}{\lambda^{3}_{T}(\textrm{O})}$
        - $\textrm{O}_{2}$ diatomikoa denez, badauzka barneko askatasun-graduak, *biraketak* eta *bibrazioak* eta partizio-funtzioa honako hau da:
        - $Z^{\textrm{O}_{2}}_{1}=\frac{V}{\lambda^{3}_{T}(\textrm{O}_{2})}\cdot Z^{\textrm{O}_{2}}_{1,\textrm{bib}}\cdot Z^{\textrm{O}_{2}}_{1,\textrm{bir}}$
            - bibrazioei dagokienez, enuntziatuak badio nolakoak diren energia-mailak

             $$Z^{\textrm{O}_{2}}_{1,\textrm{bib}}(T)=\sum\limits^{\infty}_{n=0}e^{-\frac{(n\hbar\omega+\epsilon_{0})}{k_{B}T}} \Rightarrow \boxed{Z^{\textrm{O}_{2}}_{1,\textrm{bib}}(T) = \frac{e^{-\frac{\epsilon_{0}}{k_{B}T}}}{1 - e^{-\frac{\epsilon_{0}}{k_{B}T}}}}$$

            - biraketekin lotutako *tenperatura karakteristikoa* ezaguna da, $\Theta_{\textrm{bir}}=85.4$ K, eta, kalkulatu behar den $K(T)$ konstantearen balioa $10^{3}$ K-etan kalkulatu behar denez, honako hau betetzen du: $\Theta_{\textrm{bir}}\lll T$, hortaz, biraketekin lotutako partizio-funtzioaren goi-tenperaturako limiteko adierazpena erabil daiteke, honako hau:

            $$\boxed{Z^{\textrm{O}_{2}}_{1,\textrm{bir}}(T)=4\cdot \frac{T}{\Theta_{\textrm{bir}}({\textrm{O}_{2}})}}$$

- hauxek dira azken adierazpenaK:

     $$Z^{\textrm{O}}_{1}= 4\cdot \frac{V}{\lambda^{3}_{T}(\textrm{O})}$$

     $$Z^{\textrm{O}_{2}}_{1}=\frac{V}{\lambda^{3}_{T}(\textrm{O}_{2})}\cdot \frac{e^{-\frac{\epsilon_{0}}{k_{B}T}}}{1 - e^{-\frac{\epsilon_{0}}{k_{B}T}}} \cdot 4 \frac{T}{\Theta_{\textrm{bir}}({\textrm{O}_{2}})}$$   

     $$\boxed{K_{\textrm{O}_{2}\,\rightleftharpoons\,2\,\textrm{O}}(T)=  \frac{\left(\lambda^{\textrm{O}_{2}}_{T}\right)^3}{\left(\lambda^{\textrm{O}}_{T}\right)^{6}}  \cdot  \frac{1 - e^{-\frac{\epsilon_{0}}{k_{B}T}}}{e^{-\frac{\epsilon_{0}}{k_{B}T}}} \cdot  \frac{\Theta_{\textrm{bir}}}{T}}$$   

- eta konstantearen balioa da: $K(T)\approx 1.35 \cdot 10^{7} \textrm{m}^{-3}$
