## &#9989; 25 Ariketa: kokapena 
  
## *Gainazal xurgatzailean molekulak polarizatzen dira* 
## Helburua:    

1. Multzo mikrokanonikoan kalkuluak egiten trebatzen hastea   
<br>
2. Multzo mikrokanonikotik kanonikora *pasatzen* ikastea, esangura fisikoaz konturatuz  
<br>  
  
## Ikasiko duzuna:  
   
   
----------------


Gainazal batean, $M$ tokitxo xurgatzaile ditu berak, $m$ masako $n$ partikula lotuta daude ($n$ kopurua oso-oso handia da). Tokitxo xurgatzaileek atomo bakarra har dezakete eta, hartutakoan, $AB$ molekula eratzen da. Beraz, pentsa daiteke, tokitxo xurgatzaileetan $B$ motako partikulak daudela. Eratutako molekulek badute momentu dipolarra, $\vec{p}$, bertikalki modu bakarrean orientatu daitekeena eta, horizontalki, lau modutan: $p_{z}$ eta $\pm\,p_{x}$ eta $\pm\,p_{y}$, hurrenez hurren. Lotura-energia da $\epsilon_{0}$. Demagun kanpo eremu elektrikoa ezarri dela, bertikalki: $\vec{\mathscr{\epsilon}}$. Jakina, ezarritako eremua elektrikoaren eta momentu dipolarraren arteko akoplamendu-energia da $E = - \vec{p}\cdot\vec{\Large{\varepsilon}}$. Azkenik, sistema osoaren polarizazio honako lanaren adierazpenaren bidez definituta dago: $dW = - \vec{P}\cdot d\vec{\Large{\varepsilon}}$.  
 
1. Lortu sistemaren partizio-funtzioa: $Z_{n}(T)$  
<br>  
1. Lortu potentzial kimikoaren adierazpena: $\mu(T;n)$  
<br>  
1. Ondorioztatu zenbatekoa den estaltzea: $\langle n \rangle$.  
  Kasu honetan, aurreko adierazpenetik abiatu beharko duzu estaltzea, potentzial kimikoaren funtzioan idazteko, hau da, onartuz, ezaguna dela potentzial kimikoa  
<br>  
1. Ber-lortu $\langle n \rangle$ *estaltzearen* adierazpena partizio-funtzio gran kanonikoa erabiliz  
<br>  
1. Lortu sistemaren polarizazioa  
<br>  
>Lehenengo bi galderetan sistema itxia da, xurgatutako $n$ atomo kopurua finkoa da eta, beraz, multzo kanonikoa erabili behar da.      
>
>Hurrengo bi galderetan, 3 eta 4, sistema orekan dago $A$ atomoz osatutako gasarekin. Potentzial kimikoa, beraz, emanda dago eta $\langle n \rangle$ estaltzea (*xurgatutako batez besteko atomo kopurua eta eratutako molekula kopurua*). Multzo kanonikoan lehenengo, 3.ean eta gran kanonikoaren bidez, ondoren 4.ean kalkulatu behar da.   
>
>
>Azken galderaren helburua da egiaztatzea multzo biek emaitza berbera lortzen dutela.

*Kanonikoan*:  
   
$$\boxed{\mu = (k_{B}T)\ln\left[\frac{n}{M-n}\cdot \left(4 e^{\frac{-\epsilon_{0}}{k_{B}T}} + e^{\frac{-\epsilon_{0}}{k_{B}T}}\cdot e^{\frac{+d_{z}{\large{\varepsilon_{z}}}}{k_{B}T}}\right)\right]}$$  
<br>
   <br>
- Adierazpen horretatik $\frac{n}{M}$ bakandu egin behar da:  
   
   $$\boxed{\frac{n}{M} = \frac{e^{\frac{\mu}{k_{B}T}} \left(4 e^{\frac{-\epsilon_{0}}{k_{B}T}} + e^{\frac{-\epsilon_{0}}{k_{B}T}}\cdot e^{\frac{+d_{z}{\large{\varepsilon_{z}}}}{k_{B}T}}\right)}{ e^{\frac{\mu}{k_{B}T}} \left(4 e^{\frac{-\epsilon_{0}}{k_{B}T}} + e^{\frac{-\epsilon_{0}}{k_{B}T}}\cdot e^{\frac{+d_{z}{\large{\varepsilon_{z}}}}{k_{B}T}}\right)+ 1}    }$$
   <br>
   <br>
   
*Makrokanonikoan*:  
   
   $\mathscr{Z}_{xurgatzaile}= z^{0}\cdot Z_{0} + z^{1}\cdot Z_{1}\Rightarrow \mathscr{Z}_{xurgatzaile} = 1 + z\cdot Z_{1}$     
   <br>     
   Toki xurgatzaile bakar baten *estaltzea* da batez besteko partikula kopurua: $\langle N \rangle$  
   
   Definizioz, esaterako, $\langle N \rangle = k_{B}T\left(\frac{\partial(\ln \mathscr{Z})}{\partial \mu}\right)_{T,V}$  
   <br> 
   Eta hortik lortzen da emaitza berbera.

## Ebazpena  

*Kanonikoan*:  
   
1. ${\Large{\vec{\varepsilon}}}=\vec{0} \Rightarrow Z_{1}(T)=5 e^{\frac{-\epsilon_{0}}{k_{B}T}}$  
   <br>
1. ${\Large{\vec{\varepsilon}}}\neq\vec{0} \Rightarrow Z_{1}(T)=\left(4 e^{\frac{-\epsilon_{0}}{k_{B}T}} + e^{\frac{-\epsilon_{0}}{k_{B}T}}\cdot e^{\frac{+d_{z}{\large{\varepsilon_{z}}}}{k_{B}T}}\right)$   
   <br>
1. ${\Large{\vec{\varepsilon}}}\neq\vec{0}$:  
   <br>
   $Z_{n}(T) =\left[\frac{M!}{n!(M-n)!}\right]\cdot \left(4 e^{\frac{-\epsilon_{0}}{k_{B}T}} + e^{\frac{-\epsilon_{0}}{k_{B}T}}\cdot e^{\frac{+d_{z}{\large{\varepsilon_{z}}}}{k_{B}T}}\right)^{n}$     
   <br>
   $F= -\left(k_{B}T\right)\ln Z_{n} \Rightarrow F = -\left(k_{B}T\right)\ln\left\{\left[\frac{M!}{n!(M-n)!}\right]\cdot \left(4 e^{\frac{-\epsilon_{0}}{k_{B}T}} + e^{\frac{-\epsilon_{0}}{k_{B}T}}\cdot e^{\frac{+d_{z}{\large{\varepsilon_{z}}}}{k_{B}T}}\right)^{n}\right\}$   
   <br>   
   $F=-\left(k_{B}T\right)\cdot\left\{ M\ln M - (M-n)\ln (M-n) - n\ln n - n\cdot \ln \left(4 e^{\frac{-\epsilon_{0}}{k_{B}T}} + e^{\frac{-\epsilon_{0}}{k_{B}T}}\cdot e^{\frac{+d_{z}{\large{\varepsilon_{z}}}}{k_{B}T}}\right)  \right\}$   
   <br>
   $\mu\equiv \left(\frac{\partial F}{\partial n}\right)_{T,E}$  
   <br>
   $$\boxed{\mu = (k_{B}T)\ln\left[\frac{n}{M-n}\cdot \left(4 e^{\frac{-\epsilon_{0}}{k_{B}T}} + e^{\frac{-\epsilon_{0}}{k_{B}T}}\cdot e^{\frac{+d_{z}{\large{\varepsilon_{z}}}}{k_{B}T}}\right)\right]}$$  
   <br>
   <br>
   - Adierazpen horretatik $\frac{n}{M}$ bakandu egin behar da:  
   
   $$\boxed{\frac{n}{M} = \frac{e^{\frac{\mu}{k_{B}T}} \left(4 e^{\frac{-\epsilon_{0}}{k_{B}T}} + e^{\frac{-\epsilon_{0}}{k_{B}T}}\cdot e^{\frac{+d_{z}{\large{\varepsilon_{z}}}}{k_{B}T}}\right)}{ e^{\frac{\mu}{k_{B}T}} \left(4 e^{\frac{-\epsilon_{0}}{k_{B}T}} + e^{\frac{-\epsilon_{0}}{k_{B}T}}\cdot e^{\frac{+d_{z}{\large{\varepsilon_{z}}}}{k_{B}T}}\right)+ 1}    }$$
   <br>
   <br>
   
*Makrokanonikoan*:  
   
   1. ${\Large{\vec{\varepsilon}}}\neq\vec{0}$:  
   <br>  
   $\mathscr{Z}_{xurgatzaile}= z^{0}\cdot Z_{0} + z^{1}\cdot Z_{1}\Rightarrow \mathscr{Z}_{xurgatzaile} = 1 + z\cdot Z_{1}$     
   <br>     
   ${\mathscr{Z}_{xurgatzaile}}_{M} = \left(1 + z\cdot Z_{1}\right)^{M}$   
   <br>      
   Toki xurgatzaile bakar baten *estaltzea* da batez besteko partikula kopurua: $\langle N \rangle$  
   
   Definizioz, esaterako, $\langle N \rangle = k_{B}T\left(\frac{\partial(\ln \mathscr{Z})}{\partial \mu}\right)_{T,V}$  
   <br> 
   Eta hortik lortzen da emaitza berbera.
