<div class="row alert alert-info"  role="alert">
   <br>
   <br>

Dimentsio bakarreko polimero baten eredua erabiliz deskriba daiteke banda elastikoaren elastikotasuna. Ereduaren arabera, polimeroaren molekulak muturretatatik lotuta daude, eta loturen angelua 0$^{\circ}$ edo 180$^{\circ}$ izan daiteke, eta loturetan edozein eratzeko *probabilitatea* berdina da.

1. Froga ezazu bandaren $L=2\,m\,d$ luzera ematen duten polimero-molekulen banaketa kopurua honako hau dela:   

    $$g(N,m)=\dfrac{2\,N!}{\left(\dfrac{N}{2}+m\right)!\,\left(\dfrac{N}{2}-m\right)}$$  
<br>  
1. Lortu aurreko adierazpenaren forma hurbildua $m\ll N$ denean:   

    $$g(N,m)\approx g(N,0)\, \exp\left(-\dfrac{2m^{2}}{N}\right)$$  
<br>  
1. Lortu sistemaren entropia $L$ luzeraren funtzioan, $N\gg 1$ eta $L\ll Nd$ kasuetan.  
<br>  
1. Lortu indarra eta luzeraren arteko erlazioa, $L$-ren edozein baliotarako, baina $N\gg 1$ denean.
