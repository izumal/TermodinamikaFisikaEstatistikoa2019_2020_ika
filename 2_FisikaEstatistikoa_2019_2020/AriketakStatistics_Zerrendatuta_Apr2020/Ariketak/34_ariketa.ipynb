Molekula diatomikoak bibratzen duenean, haren inertzia-momentuak badauka, neurri txiki batean bada ere, bibrazioarekiko mendekotasunik. Jakina, hori horrela, bibrazioak eta biraketak ez dira guztiz independenteak. Demagun honako hau dela bibrazioekin eta biraketekin lotutako energia-espektroa:  

$$ 
E_{n,l}=
\left\{\hbar\,\omega(n+\dfrac{1}{2})\right\}_{bib}+
\left\{\dfrac{\hbar^{2}}{2I}l(l+1)\right\}_{bir}+
\left\{\alpha\,l(l+1)\,(n+\dfrac{1}{2})\right\}_{bib-bir}
$$  

adierazpen horretan, lehen eta bigarren atalak dira bibrazio-energia eta biraketa-energia, hurrenez hurren eta hirugarrena, bibrazioen eta biraketen arteko loturatik datorrena. Adierazpen horretan agertzen diren konstanteek honelako erlazio hauek betetzen dituzte:   

$$\hbar\omega \ggg \dfrac{\hbar^{2}}{2I} \ggg  \alpha$$   


1. Lor ezazu horrelako molekulez osatutako gasari dagokion barne-energiaren adierazpena, $T$ tenperaturak honako hau betetzen badu:   

$$\hbar\omega>k_{B}T\ggg \dfrac{\hbar^{2}}{2I}$$.
