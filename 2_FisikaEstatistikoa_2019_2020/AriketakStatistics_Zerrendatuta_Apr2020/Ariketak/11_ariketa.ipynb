Artikulu honetan daude proposatuta: https://arxiv.org/abs/1207.2121  
<br>

Aztertu beharreko sistema honako hau da: bi dimentsioko sarean edozein konfiguraziotan ager daitekeen $a$ luzerako $N$ monomeroz osatutako homopolimeroa.
Polimeroaren muturretako bat finkoa da eta erreferentzi-sistemaren jatorritzat dugu. Gainera, zirkulu irekiaren bidez adierazita dago, ikus irudia.
Polimetroaren beste muturrari $\tau$ tentsioak eragiten dio: izan ere, beti eragiten dio $OX$ ardatzak definitutako norabidean.   

$L_{x}$ da polimeroaren muturren arteko distantzia $OX$ ardatzean neurtuta. Berebat $L_{y}$ da polimetroaren muturren arteko distantzia $OY$ ardatzean neurtuta.   

1. 1 aldaera: **monomeroak gainazalean edozein modutan orientaturik**  

    Monomeroak koka daitezke $OX$ ardatzean noranzko batean zein bestean, hots $x$-positiboetara zein negatiboetara. Horrelako egoerekin lotutako lanak (lan elastikoak) honako hauek dira: $-\tau\,a$ eta $\tau\,a$, hurrenez hurren.
Gainera, monomeroak koka daitezke $OX$ ardatzarekiko perpendikular $+y$ zein $-y$ noranzkoetan orientaturik. Horrelako egoerei $\epsilon$ energia esleituko diegu

<br>  
2. 2 aldaera: **monomeroak gainazalean, $OX$ ardatzean, soilik $+x$ noranzkoan**  

    Kasu honetan monomeroak $+x$ noranzkoan baino ezin dira kokatu, hots, $-x$ noranzkoa debekatuta dago. *Horma zeharkaezinak* adierazten du baldintza hori, ikus irudia. $OX$ ardatzean balio positiboak baino ez dira onartzen.

<br>      
3. 3 aldaera: **monomeroak gainazalean, $OX$ eta $OY$ ardatzetan, soilik $+x$ eta $+y$ noranzkoetan**  

    Monomeroak $+x$ noranzkoan baino ezin dira kokatu eta, gainera, $+y$ positboan baino ezin dira agertu $OY$ ardatzean. Bi ardatzetan balio positiboak baino ez dira onartzen.
