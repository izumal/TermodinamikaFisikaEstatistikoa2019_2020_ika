Multzo mikrokanonikoaren eta kanonikoaren erabileen arteko aldea berriro ere azpimarratzeko, ebatz dezagun honako ariketa hau. Polimero baten eredu simplea bi dimentsiotan.  

Bi dimentsioko sarean, sare-puntuetan, eratzen den ibilibidea da polimeroa. Edozein sare-puntutan, polimeroak aurrera egin dezake, 1 aukera aldameneko irudian, edo, bestela, ezkerretara zein eskuinetara, perpendikularki orientatuta kokatuta dagoen orientazioekiko, 2 eta 3 aukerak irudian. Ezkerretara zein eskuinetara egiten duen bakoitzean, $\epsilon$ makurtze-energia gastatu behar du. Beraz, polimeroaren eiteei energia bana dagoekie: makurtze-energia bider ezkerretara eta eskuinetara egindako kopuruen batura.  
    
   ![AriketakIrudiak/3Ariketa5azkena, width=150 frac=0.8]()   

Demagun polimeroaren hasierako puntua sare-puntu jakin batean dagoela eta polimeroa eratzen $N+1$ segmentu daudela. Orduan, polimeroaren eite posibleak dira sistemaren (mikro)egoera estatistikoak.   
 
1. Zenbat eite posible dago $E=m\,\epsilon$ energiarekin? $\hspace{0.4cm}m$ balioa da $0\le m \le N$ tartekoa.    
<br>      
1. Zenbatekoa da sistemaren $S=S(E, N)$ entropia?  
<br>  
1. Kalkulatu sistemaren tenperatura, $E$ eta $N$ parametroen funtzioan.   
<br>  
1. Kalkulatu sistemaren $E$ energia, $T$ eta $N$ parametroen funtzioan.   
<br>  
1. Kalkulatu $C_{N}$, luzera konstanteko bero-ahalmena,  $T$ eta $N$ parametroen funtzioan.
