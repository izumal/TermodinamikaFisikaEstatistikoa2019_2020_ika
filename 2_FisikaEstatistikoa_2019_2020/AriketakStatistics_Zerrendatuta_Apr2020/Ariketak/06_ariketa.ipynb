   <h2> Ebazpena </h2>

**Multzo mikrokanonikoan**, dena konstante da, $E, N, V$, fisikoki sistema unibertsoa da, erabat isolatutako sistema.

1. Idatz itzazu itxitura baldintzak:   

   * partikula kopuru osoarena  
   * energia osoarena, har ezazu kontuan energietako bat negatiboa dela   
   
1. Erreparatu honi buruz: $N$ konstantea denez, maila bakoitzean dagoen partikula kopuruari dagokionez, aldagai bakarra dago: $n_{0}$ edo $n_{1}$.  

1. Egin honako aldagai-aldaketa. Definitu energia-mailetako partikula-frakzioa, i.e. $n_{i}^{'}\equiv\dfrac{n_{i}}{N}, i=1, 2$; eta energiaren itxitura-baldintzako ekuazioatik ondorioztatzen den $x$ aldagaia.  

1. Berridatzi energia-mailetako partikula-frakzioak definitu berri duzun $x$ aldagi adimentsionala erabiliz: $x\equiv\dfrac{E}{N\,\epsilon}$.  

1. Lortu $\Omega$ mikroegoeren kopurua, energia-mailetan dagaoen partikula kopuruak erabilita eta aplikatu *Stirling-en* hurbilketa. Honako hau lortu beharko zenuke:  

$$S = k_{B}\, \left[ N\ln N -n_{+}\ln n_{+} -n_{-}\ln n_{-} \right]$$    

----------  


**Multzo kanonikoaren kasuan**, abiapuntua termodinamika ondorioztatzeko da ***partizio-funtzioa***. Horixe da beraz, oraingo honetan lortu beharrekoa. Dena dela, ezer egin baino lehen, pentsatu apurtxo bat. Kontuara zaitez, sistemaren partikula osatzaileak identikoak izateaz gain, elkarrekintzarik gabekoak direla. Horrek esan nahi du, batari *gertatzen zaiona* beste edozeini gertatzen zaiola, denak aldi berean tangan sartuta egonik. Beraz, partikula baten kasua haintzakotzat hartuz, $N$ partikulez osatutako multzoari dagokiona lor dezakezu.  

1. Lortu partikula bakar bati dagokion partizio-funtzioa.  

   Jakina, definizioa ez da aldatzen, beraz, argi dago bi atal baino ez duela izango.  
   
1. Idatzi $N$ partikulez osatutako multzoari dagokion partizio-funtzioa, honako hau: $Q_{N}(V, T) = \sum_{r} \exp^{-\dfrac{E_{r}}{k_{B}\,T}}$  

1. Adierazpen horretan, $E_{r}$ da multzoari (ez multzo estatistikoari, kontuz) dagokion energia-maila posibleetako bat. Beraz, lortu berri duzun adierazpenetik honako hau idatzi beharko zenuke:   

    $$Q_{N}(V, T) = \left(Q_{1}(V, T)\right)^{N}$$  

1. Hau da, multzoaren partizio-funtzioa da $N$ aldiz partikula bakarrarena.  

1. Horren bidez, lortu multzo makrokanikoaren formalismoan oinarrizkoa den potentzial termodinamikoa, hots, $F(T, V, N)$, Helmholtz-en energia askea:   

    $$F(T, V, N) = - k_{B}T\,N\, \ln[Q_{1}(V, T)]$$  

1. Kontuan izan, entropiaz galdetu duela enuntziatua. Lortu entropia, $F$ potentzialetik abiatuta:   

    $$S = - \left(\dfrac{\partial F}{\partial T}\right)_{V, N}$$  

1. Aurreko puntuan lortutako adierazpena honako hau da, partikulako $S$ entropia, hain zuzen ere:   

    $$ \dfrac{S}{N} = k_{b} \left\{\ln \left[ e^{\dfrac{\epsilon}{k_{B}T}} + e^{\dfrac{\epsilon}{k_{B}T}} \right] + \dfrac{\epsilon}{k_{B}T}\, \left\{\dfrac{e^{\dfrac{\epsilon}{k_{B}T}} - e^{\dfrac{\epsilon}{k_{B}T}}}{e^{\dfrac{\epsilon}{k_{B}T}} + e^{\dfrac{\epsilon}{k_{B}T}}}\right\}\right\}$$  

---------  

**Parentesia**  

1. Berreskuratu mikrokanonikoaren formalismoan lortu duzun entropiaren adierazpena: $$ S = Nk_{B}\,\left[ \ln 2 - \frac{1}{2} (1+x)\ln (1+x) -\frac{1}{2}(1-x)\ln (1-x) \right] $$  

1. *Sartu* fisika multzo mikrokanonikoan, $\dfrac{1}{T}$ funtzioa kalkulatuz  

1. Kontuan izan honako hau:  

    $$ \dfrac{1}{T} = \left( \dfrac{\partial S}{\partial E} \right){V, N} = \left( \dfrac{\partial S}{\partial x} \right)\, \left(\dfrac{\partial x}{\partial E} \right){V,N}$$  
   
1. Frogatu hauxe betetzen dela:   

    $$ \dfrac{\epsilon}{k_{B}T} = \frac{1}{2}\,\ln\left(\dfrac{1-x}{1+x}\right)$$  

1. Erabili berdintza hori, eta lortu beharko duzun besteren bat, 6. ataleko partikulako entropiari dagokion adierazpen berria lortzeko.  

1. Pentsatu nola eraiki behar den multzoari dagokion energia partikulei dagozkien energiak erabilita.
