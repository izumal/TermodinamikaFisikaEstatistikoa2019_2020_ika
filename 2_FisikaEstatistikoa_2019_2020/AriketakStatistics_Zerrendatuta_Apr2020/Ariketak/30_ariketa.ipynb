**Ariketa hau da aurrekoaren bertsioa**   

Gas ideal klasikoa $V$ bolumeneko tangan sartu dugu. Tangaren hormek gasaren partikulak xurga ditzakete, $N_{0}$ sare-puntu xurgatzaile daude hormetan, hain zuzen. Bi partikula gehien jota xurga ditzakete sare-puntuek. Eta xurgatutako partikulen energia $-\epsilon$ da. Gasaren partikula kopurua $N$ da, finkoa bera eta, $2N_{0}$ baino handiagoa.   

1. Erabili multzo kanonikoa horma-xurgatzaileekin ukipenean dagoen gasaren egoera-ekuazioa lortzeko.   
<br>    
1. Berebat, lortu xurgatutako partikulen batez bestekoa, $T\to 0$ eta $T\to\infty$ limiteetan.
