# 1. Ariketa

## Solido ez perfektuak (1): *Frenekel-en* akatsdun solidoak

<br>
<br>

Aztertu beharreko sistema da $N$ atomo identikoek eratzen duten solidoa, hiru dimentsioko banaketa erregularra. Atomo horietako $n$ ($1\gg n\gg N$) dauden sare-puntuetatik kendu eta sare-puntuen artekoetan ezartzen badira, *Frenkel-en* motako akatsadun solido eratuko da, $n$ akatsdun solidoa. Solidoan dauden sare-puntuen artekoen kopurua, $N'$, sare-puntuen kopuruaren ($N$) ordenakoa da.  

Demagun $\epsilon$ dela beharrezkoa den energia *Frenekel-en* motako akatsa sorrarazteko, hots, beharrezkoa den energia sare-puntu batean dagoen atomoa *hutsune* (sare-puntuen artekoa) batera pasarazteko.  

>Frogatu honako adierazpen hurbildua egokia dela oreka-egoeran, $\epsilon \gg k_{B}T$ betetzen duen $T$ tenperaturan,  

><br>
>$$\boxed{ n \approx \sqrt{NN'} \exp^{-\frac{\epsilon}{2k_{B}T}} }$$



<br>
<br>
## Solido ez perfektuak (2): *Schottky-ren* akatsdun solidoak

<br>
<br>


Aztertu beharreko sistema da $N$ atomo identikoek eratzen duten solidoa, hiru dimentsioko banaketa erregularra. Atomo horietako $n$ ($1\gg n\gg N$) dauden sare-puntuetatik kendu eta solidoaren gainazaleko sare-puntuetan ezartzen badira, *Schottky-ren* motako akatsadun solido eratuko da, $n$ akatsdun solidoa.    

Demagun $\epsilon$ dela beharrezkoa den energia *Schottky-ren* motako akatsa sorrarazteko, hots, beharrezkoa den energia (solidoaren barneko) sare-puntu batean dagoen atomoa gainazaleko sare-puntu batera pasarazteko.  

>Frogatu honako adierazpen hurbildua egokia dela oreka-egoeran, $\epsilon \gg k_{B}T$ betetzen duen $T$ tenperaturan,  
><br>
>$$\boxed{n \approx N \exp^{-\frac{\epsilon}{k_{B}T}} }$$  
<br>
> Onartu akatsak sorrarazten direnean, gertatzen den bolumen-aldaketa baztergarria dela.

# 2. Ariketa
## Solido ez perfektuak (3): sare-puntu xurgatzaileak dituen gainazaldun solidoak

<br>
<br>


Solido baten gainazalean $N_{0}$ sare-puntu xurgatzaile dago eta horietan guztira xurgatutako gas-molekulen kopurua da $N$.  

Lortu xurgatutako gas-molekulen potentzial kimikoa.  

Baztertu xurgatutako molekulen arteko elkarrekintzak.   

**`Bonus:`**  
Proposatu ariketa ebazteko *multzo estatatistiko kanonikoaren formalismoa* erabiltzen ez duen beste moduren bat.

# 3. Ariketa
## Kate molekularra (1): luzera ezberdineko 2 unitatezkokate molekularra

<br>
<br>


Dimentsio bakarreko kate molekularra eratuz $N$ unitate dago. Unitatetak bi egoeretan egon daitezke, $a$ eta $b$ eta, egoeretan, unitateen luzerak eta energiak dira $L_{a}$ eta $L_{b}$ eta $\epsilon_{a}$ eta $\epsilon_{b}$, hurrenez hurren.  

>Ondorioztatu dimentsio bakarreko katearen muturretan ezarri beharreko $\tau$ tentsioaren eta horrek mantentzen duen $L$ luzeraren arteko lotura. 
><br>  
>Erabili ***tentsio konstantepeko multzo kanonikoa***.  




`Nota:` Irudiko $X$ da enuntziatuko $\tau$ eta $a$ eta $b$, $L_{a}$ eta $L_{b}$, hurrenez hurren.

# 4. Ariketa
## Kate molekularra (2): luzera bakarreko eta biraketa askeko loturak dituzten unitatezko kate molekularra

<br>
<br>



Kate molekular bat eratzen $a$ luzerako $N$ unitate daude. Unitateen arteko loturak bereziak dira eta ahalbidetzen dute lotzen dituzten unitateen orietazioak edozein izatea, hots, unitateek era askean bira dezakete loturetan.   

>Ondorioztatu hiru dimentsioko katearen muturretan ezarri beharreko $\tau$ tentsioaren eta horrek mantentzen duen $L$ luzeraren arteko lotura. 
><br>  
>Erabili ***tentsio konstantepeko multzo kanonikoa***.
>
>Onartu bibrazio-energiak eta beste edozein energia motak ez daukala molekularen formarekiko mendekotasunik.   



`Nota:` Irudiko $X$ da enuntziatuko $\tau$.
