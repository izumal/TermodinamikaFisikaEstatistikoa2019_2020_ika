## Multzo Estatistiko Mikrokanonikoa: *...fisika sartzea*    
<br>  
<br>  
01. [01 Ariketa](https://nbviewer.jupyter.org/github/jmigartua/TermodinamikaFisikaEstatistikoa2018_2019_ika/blob/master/2_FisikaEstatistikoa_2018_2019/AriketakStatistics_Zerrendatuta_Apr2019/Ariketak/01_ariketa.ipynb), **Solidoak: Dimentsio bakarreko akatsdun solidoa**    
<br>
Izadian solido perfektuak ez dira exisititzen, nahiz eta normalean egiten dugun teorian, horrelakoak izaten diren. Kasu honetan, *akatsak* dituen solido bat aztertzeko lehenengo urratsak emango dira. *Puntu-akatsak* izango dira aztertuko ditugunak, hau da, partikula osatzaile bakar batekin lotutakoak.
<br>
Demangun solidoa dimentsio bakerrekoa dela, eta atomo bakarrekoa, monoatomikoa, partikula puntualak haiek. Onartuko da solido horren gelaxka-unitatearen dimentsioa $a$ luzera dela. Beraz, solido perfektua izanik, dimentsio bakerrean, $a$ distantziara kokatuz, hedatzen diren puntu materialen segida infinitua dugu. Puntu material horien artean ez dago elkarrekintzarik. Orduan, solidoaren abstrakzioa egin daiteke eta honako hau esan: solidoa da puntu materialez osatutako sare infinitua, dimentsio bakarreko $a$ luzerako gelaxka-unitatea era infinituan errepikatuz sorrarazitakoa. Egoera horretan, sistema perfektu horren energia jatorriko energia da, zero.
<br>
Horrelako solidoan egon daitezken puntu-akatsak dira sare-puntuetatik kanpoko atomoak agertzea, esaterako. Akatsek solidoaren energia handituko dute, jatorritik ateraz, beraz. Jakina, ohiko sare-puntu batetik kanpo atomo bat dagoenean, alboetara dituen ohiko sare-puntuetan dauden atomoak ikusiko ditu, izango du elkarrekintza. Elkarrekintza hori adierazteko $\epsilon$ energia erabiliko da. Hau da, sare-puntuetako atmoekin lotutako energia da $\epsilon_{S-P} = 0$ eta sare-puntuetatik kanpokoekin (ez-ohiko sare-puntuak dira horiek) lotutakoa, $\epsilon_{eS-P} = \epsilon$.
