## **Termodinamikaren 0. Printzipioa**


## Ariketak  (1-5)

Ondoren dauden ariketak termodinamikaren Zero Printzipioarekin lotutako ariketak dira. *Zemansky* liburutik ateratakoak dira, esplizituki adierazita dagoen moduan.

**Ariketen helburua**: tenperatura-funtzioaren kontzeptuaz jabetzea eta oreka termikoarekin lotutako adibideak ebazten trebatzea.

Ikusiko duzuenez, elkarren arteko ukipen termikoan esarriko diren sistemak mota berekoak izan daitezke (hots, haien deskribapenean parte hartzen duten aldagaiak mota berekoak dira, askatasun-gradu bereko sistemak izango baitira) edo ez. Azken kasuan, jakina, sistemen deskribapenean parte hartzen dituzten aldagaiak mota desberdinekoak dira, baina edozein kasutan, sistemek izango dute esleituta ***tenperatura-funtzioa***. Oreka termikoan dauden sistemen tenperatura-funtzioek balio beberbera izango dute, oreka termiko, egoera fisiko, berean baitaude. Oreka termikoa adierazteko, tenperatura-funtzioak erabilita, **berdintzak** (ekuazio matematikoak) erabiltzen dira. Horietan dautza ariketak.

## Tenperatura-funtzioa sistema hidrostatiko ezberdinen kasuan

Ariketa honetan, *sistema hidrostatikoak* dira elkarren arteko oreka termikoan dauden sistemak. Horien deskribapenean parte hartzen dituzten aldagaien sorta honako hau da (). Sistemak desberdintzeko aldagaiak desberdintzen dira.

Oraindik ez dira sistema hidrostatikoak definitu, baina, ariketa hauetarako jakin beharreko gauza bakarra da horien kasuan deskribapenean parte hartzen duten aldagaiak direla honako sortakoak: $p, T (\theta), V$.

   
>### Helburua:  
>   
>   1. *Oreka termikoa* kontzeptuaz adibideetatik jabetzea 
>   
>   2. Termodinamikaren 0. Printzipioa erabiltzen trebatzea  
>   
>   
>###  Ikasiko duzuna:    
>
>   1. Zero printzipiotik ondorioztatutko duzu *tenperatura-funtzioa*.   
>      Horretarako egin beharreko gauza bakarra da oreka termikoa defintzen duen funtzio inplizitua funtzio ezplituetan banatzea, hots, sistema bakoitzari dagokiona bakantzea
>   <br>  
>   2. Nahiz eta sistema mota berdina izan, tenperatura-funtzioa ezberdina izan daiteke.   
>      Sistem mota berdina izanik baina sistemak, sistema konkretuak ezberdinak badira, orduan, tenperatura-funtzioak berdinak dira  
>   <br>       
>   3. Tenperatura-funtzioak, ezberdinak izanik ere, balio berbera dute.  


   
---------------------------

### Enuntziatua:  


$A$, $B$ eta $C$ gasak dira, ($p$, $V$), ($p'$, $V'$), ($p''$, $V''$) parametro-bikoteen bidez ezaugarritutakoak.

$A$ eta $C$ oreka termikoan dauden kasuan, honako ekuazio hau betetzen da:
$$pV - nbp - p'' V''  = 0$$
$B$ eta $C$ oreka termikoan dauden kasuan, ordea, betetzen den ekuazioa honako hau da:
$$ p\,'V' - p''\,V''  + \frac{n\,B'\,p''\,V''}{V'} = 0 $$
adierazpen horretan $n$, $B'$ eta $b$ konstanteak dira.

- Zein dira oreka termikoan elkarren berdinak diren funtzioak, non bakoitza $t$ tenperatura enpirikoaren berdina den?

$$ 4\,\pi n\,R\,C_C\,H - M\,p\,V = 0 $$
$B$ eta $C$ oreka termikoan dauden kasuan, berriz, ondokoa:
$$ n\,R\,\Theta\,M' + 4\,\pi\,n\,R\,C'_C\,H'  - M'\,p\,V = 0 $$
adierazpen horretan $n, R, C_c, C_{c^{'}}$ eta $\Theta$ konstanteak dira.

## Bolumen konstanteko gas-termometroa

Ariketa honetan gogorarazi nahi da bolumen konstanteko gas-termometroaren (bkgt) erabilera eta gas idealaren tenperatura-eskalaren definizioa. Nahiz eta bkgt-aren tanga egoera termiko berean dagoen egoerarekin ukipen termikoan ezarri, bolumenean bertan sartutako gas kantitatea ezberdina bada, lortuko den tenperaturaren balioa ezberdina izango da. Haatik, gas idealaren limetean, hots, sartutako gas kantitatea oso-oso txikia denean, tenperaturaren balio bakarra lortzen da. Kasu honetan, gas aldatuko bagenu ere bai.  




>## Helburua:

>##  Ikasiko duzuna:

>      Baldintza esperimentalen araberakoak dira balaio horiek.

   
---------------------------

### Enuntziatua:  


Bolumen konstanteko gas-termometroaren gordailuan dagoen gasaren presioa adierazi dute beheko taulako lehenengo lerroko zenbakiek, gordailua uraren puntu hirukoitzean dagoen kasuan. Beheko lerroak, aldiz, gordailuaren presioaren irakurketak adierazi ditu, termometroa beste egoera termikoan dagoenean. Zutabeetan, jakina, boumen-konstanteko gas-termometroan sartutako gas berberaren kantita ezberdinak kokatuta neurtutako presio-balioak daude.

- Lortu substantziari dagokion $\theta$ tenperatura,  gas idealaren tenperatura-eskalan.

| egoera termikoa | $m_1$ | $m_2$ | $m_3$ | $m_4$ |

| p (mm Hg) PH   | 1000.0 |	750.0  | 500.0  | 250.0  |

- **Laguntza**: irudikatu puntuak $\theta\,/p_{PH}$, doitu erregresio lineala erabiliz eta doiketaren *jatorriko ordenatua* izango da oreka termikoari dagokion tenperaturaren balioa.

- Gogoratu honako hau dela tenperaturaren balio gas idealaren tenperatura-eskalan:
$$\theta = 275.16\,\times\,\lim_{PH\to 0}\frac{p}{p_{PH}} $$

Beharrezkoak direnak inportatu

```python
import matplotlib.pyplot as plt
import numpy as np
```

Tenperaturak kalkulatzeko funtzioen definizioak

```python
def tenperatura_1_4(erresistentzia):
    a = -1.16
    b = 0.675
    return np.log10(erresistentzia) / (a + b * np.log10(erresistentzia))**2.0

def tenperatura_1_5(erresistentzia):
    a = 4.697
    b = -3.917
    return 10**( (np.log10(erresistentzia) - a) / b )
```

Tenperaturen balioak lortu

```python
print round(tenperatura_1_4(1000), 3)
print round(tenperatura_1_5(218), 3)
Irudiak egiteko erresistentzien ardatzeko balioak Kalkulatu

erresistentzia_1_4 = np.linspace(1000, 30000, 101)
erresistentzia_1_5 = np.linspace(200, 30000, 101)

tenperaturak_1_4 = []
for r in erresistentzia_1_4:
    tenperaturak_1_4.append(tenperatura_1_4(r))

tenperaturak_1_5 = []
for r in erresistentzia_1_5:
    tenperaturak_1_5.append(tenperatura_1_5(r))
```

Irudikatu

```python
fig_1_4_5, axes = plt.subplots(nrows=1, ncols=1, figsize=(25,15), )
ax0 = axes

fig_1_4_5.text(0.35, 0.94, r'1.4 eta 1.5 ariketen emaitzak', 
               size=30, va='center', rotation='horizontal')

ax0.tick_params(axis='y', labelsize=25)
ax0.tick_params(axis='x', labelsize=25)

rc('text', usetex=True)
rc('font', family='serif')

ax0.axhline(y=4, xmin=0, xmax=1, lw=4, alpha=0.25, c='k')
ax0.axvline(x=218, ymin=0, ymax=1*4./4.25, lw=4, alpha=0.25, c='b')
ax0.axvline(x=1000, ymin=0, ymax=1*4./4.25, lw=4, alpha=0.25, c='r')

ax0.set_xlabel(r'Erresistentzia ($\Omega$)', size=30)
ax0.set_ylabel(r'Tenperatura (K)', size=30)

ax0.plot(erresistentzia_1_4, tenperaturak_1_4, 
color='red', lw=8, alpha=0.5, label='1.4 Ariketa');
ax0.scatter(1000, tenperatura_1_4(1000), color='k', s=200);

ax0.plot(erresistentzia_1_5, tenperaturak_1_5, 
color='blue', lw=8, alpha=0.5,  label='1.5 Ariketa');
ax0.scatter(218, tenperatura_1_5(218), color='k', s=200);

ax0.legend(loc="upper right" , prop={'size': 35});
```
