<br>

<br>

Idatz dezadan $L$, $T$ eta $\tau$ aldagaien funtzioan, $L=L(T, \tau)$ eta kalkula dezadan existitzen den funtzio horren, aurkitu nahi dudan egoera-ekuazioaren, diferentziala, honako hau:  
    
 $$ dL = {\left(\frac{\partial L}{\partial T}\right)}_{\tau}dT + {\left(\frac{\partial L}{\partial\tau}\right)}_{T}d\tau $$  
 
Ezer egin baino lehen, erraztuko ditut emandako datuak, definizioekin alderatuta:  
<br>
$$ \alpha =  \frac{1}{L}{\left(\frac{\partial L}{\partial T}\right)}_{\tau} \Rightarrow \frac{1}{L}{\left(\frac{\partial L}{\partial T}\right)}_{\tau} =  -2\,\left(\frac{(L-L_{0})}{L\,T}\right)$$  
<br>  
Beraz, honako hau idatz dezaket:  
<br>  
  $$ {\left(\frac{\partial L}{\partial T}\right)}_{\tau} = -2\,\left(\frac{(L-L_{0})}{T}\right) $$
<br>  
<br>
$$ Y \equiv \frac{L}{A}{\left(\frac{\partial\tau}{\partial L}\right)}_{T}\Rightarrow \frac{L}{A}{\left(\frac{\partial\tau}{\partial L}\right)}_{T} = \frac{L}{A}\,aT^{2}\Rightarrow {\left(\frac{\partial\tau}{\partial L}\right)}_{T} = aT^{2}$$  
<br>  
  Beraz, honako hau eidatz dezaket:   
<br>  
  $$ \frac{1}{\left(\frac{\partial\tau}{\partial L}\right)_{T}} = {\left(\frac{\partial L}{\partial \tau}\right)}_{T} = \frac{1}{aT^{2}} $$  
<br>
Orduan, adierazpen orokorrena lortutakoak ordezkatua, hauxe daukat:  
<br>
$$  dL =  -2\,\left(\frac{(L-L_{0})}{T}\right) dT + aT^{2} d\tau $$  
<br>
Horrela idatzita, ezin da integratu, beraz, moldatu egingo dut, honako hau idatziz:  
<br>
$$  dL + 2\,\left(\frac{(L-L_{0})}{T}\right) dT = aT^{2} d\tau $$  
<br>
$$  aT^{2}\,dL + aT^{2}\,2\,\left(\frac{(L-L_{0})}{T}\right) dT = d\tau \Rightarrow d\tau = aT^{2}\,dL + aT^{2}\,2\,\left(\frac{(L-L_{0})}{T}\right) dT \rightarrow \boxed{d\tau = aT^{2}\,dL + 2aT\,(L-L_{0}) dT}$$  
<br>
Azken hori erraz da integratzen, lehenik egiaztatuta ahal dela; hots, lehenik, diferentzial zehatza dela egiaztatu behar da. Horretarako, bigarren ordenako deribatu partzila gurutzatuek berdin izan behar dute:  
<br>

$$ \left(\frac{\partial^{2} \tau}{\partial L\partial T}\right) = \left(\frac{\partial^{2} \tau}{\partial T\partial L}\right)$$  
<br>
Honako hau, beraz:  
<br>
$$ \left(\frac{\partial\left[2aT\,(L-L_{0})\right]}{\partial L}\right)_{T} = \left(\frac{\partial \left[aT^{2}\right]}{\partial T}\right)_{L} $$
<br>
Hala denez, orduan diferentzial zehatza da eta integratu daiteke. Honako hau da integrala, egoera-ekuazioa, beraz:   
<br>

$$ \boxed{\tau = cT^{2}\,(L-L_{0})} $$

Lehen printzipioaren adierazpen diferentziala:  

$$ \delta Q = dU -\delta W $$  
Sistema magnetikoetan lanaren definizioa aurreko adierazpen orkorrean ordezkatuz honako hau lortzen da:  
$$ \delta Q = dU - HdM $$  
Kontuan hartuko dut lortu beharreko adierazpena prozesu adibatiko kuasiestatikoei dagokiena dela, beraz, $\delta Q = 0$ beteko da, honako hau da integratu beharreko ekuazio diferentziala:  
$$ \boxed{0 =dU - HdM} $$  
Integratu ahal izateko finkatu behar dira aldagi termodinamiko egokiak, edozein. Aukera egokienetakoa da $(T, M)$. Horren kasuan, honako hau da barne-energiaren adierazpen diferentziala:  
$$ dU = \left(\frac{\partial U}{\partial T}\right)_{M}dT + \left(\frac{\partial U}{\partial M}\right)_{T}dM $$  
Hauxe da adierapena koefiziente esperimentalen funtzioan:  
$$ \boxed{dU = C_M dT + \left[\frac{C_{Y} - C_{M}}{\left(\frac{\partial M}{\partial T}\right)_{H}} + H\right]dM}$$  
Eta adierazpen hori prozesuari dagokion ekuazio diferentzialean *sartuz*:  

$$0 = C_{M} dT + \left[\frac{C_{Y} - C_{M}}{\left(\frac{\partial M}{\partial T}\right)_{H}} + H\right]dM - HdM$$  

Honako hau, hortaz:  

$$ C_{M} dT = - \left[\frac{C_{Y} - C_{M}}{\left(\frac{\partial M}{\partial T}\right)_{H}}\right]dM$$   
Adierazpena errazteko, kontuan hartuko dut *Mayer-en* erlazioa, sistema magnetikoen kasuan:  

$$ C_{Y} - C_{X} = - T \left[\left(\frac{\partial M}{\partial T}\right)_{H}\right]^{2}\,\left(\frac{\partial (-H)}{\partial M}\right)_{T} $$  
Behar dena $\frac{C_{Y} - C_{M}}{\left(\frac{\partial M}{\partial T}\right)_{H}}$ denez, honako hau da emaitza:  

$$ \left[\frac{C_{Y} - C_{M}}{\left(\frac{\partial M}{\partial T}\right)_{H}}\right]=  T \left(\frac{\partial M}{\partial T}\right)_{H}\,\left(\frac{\partial H}{\partial M}\right)_{T} = T \left(\frac{\partial M}{\partial T}\right)_{H}\,\frac{1}{\left(\frac{\partial M}{\partial H}\right)_{T}} = - T \left(\frac{\partial H}{\partial T}\right)_{M} $$

Azkenik, integratu beharrekoa hauxe da:   

$$ \boxed{C_{M} dT =  T \left(\frac{\partial H}{\partial T}\right)_{M} dM }$$  
Egoera-ekuazio magnetikoa erabilita $\left(\frac{\partial H}{\partial T}\right)_{M} = \frac{M}{c}$. Orduan,  

$$ C_{M} dT =  T \frac{M}{c} dM \Rightarrow \boxed{\frac{C_{M}c}{T}dT = MdM}$$  
Eta horren integrala honako hau da:  

$$ \boxed{T^{cC_{M}}e^{\frac{1}{2}M^{2}} = C}$$
<br>
Azken adierazpen horretan, erabili egoera-ekuazio magnetikoa eta ($H, M$) aldagai-bikotean idazteko:  
<br>
$$ \boxed{ C =\left(\frac{M}{H}\right)^{cC_{M}}\,e^{\frac{1}{2}M^{2}}  }  $$

  <h4><i class="fa fa-pencil"></i> 4. Ariketa: Lehen printzipioaren erabilera</h4>
</div>
<br>
<br>
Horma adiabatiko higikor batek bi zatitan banatu du zilindro bat. Zilindroko ezkerreko gunea horma diatermoz inguratuta dago; eta eskuinekoa, aldiz, horma adiabatikoz.  

Eskuineko gunean $C_{V}$ konstanteko eta honako egoera-ekuazio mekanikoa duen sistemaren mol bakarra dago:   
<br>
$$p(V-b) = RT$$   

Ezkerreko gunean dago *van der Waals-en* fluidoaren beste mol bakarra.

Hasiera batean, guneetako azpisistema bien hasierako egoera berdina da: ($p_{0}$, $T_{0}$).  

Ezkerreko guneari beroa eman zaio, era kuasiestatikoan, haren $p$-ren balioa $2\,p_{0}$ izan arte.  

Ezkerreko guneko gasaren barne-energiaren aldaketa ezaguna da: $\Delta U^{ez}$.  
<br>
Lortu:  
1. **Eskuineko guneko gasaren barne-energiaren aldaketa.**  

2. **Ezkerreko guneko gasak xurgatu duen beroa.**

Ariketa hau dagoeneko eginda dago ariketa-eskola batean. Beraz, nola egin behar den ezaguna da. Oraingo honetan, ariketaren *helburua* beste bat da: konturatzea, alde batetik, ezkerreko gunean dagoen gasari buruzko informazioa ez dela beharrezkoa, gertatzen zaion dena eskuinean dagoen gasari gertatzen zaiona ezaguta lor daiteke eta, bestetik, konturatzea eskuinekoan dagoen gasa ideala ez denez, ez dela ezaguna zer barne-egitura (zer motatako mendekotasunak dauzkan aldagai termodinamikoekiko) duen barne-energiak eta nolakoak diren horrelako gasari dagozkion prozesu adiabatiko kuasiestatikoen forma (edozein diagramatan).  
<br>
**Ez dira behar *van der Waal-en* fluidoari dagozkion egoera-ekuazioak, prozesuaren ondoriozko barne-energiaren aldaketa datua da.**  
<br>
Dagoeneko, esan bezala, ariketa eginda dago:  
<div class="well" markdown="1">
 1. Gertatzen den prozesua kuasiestatikoa da, ezkerreko guneari era kuasiestatikoan ematen baitzaio beroa.  
<br>
 2. Eskuineko gasak jasaten duen prozesua adibatiko kuasiestatikoa da.   
<br>
 3. Bi guneen arteko bereizte-hormaren adiabatikotasunak termikoki isotzen ditu bi guneetako gasak.  
<br>
 4. Baina bereizte-horma hori higikor izateak eta prozesua kuasiestatiko izateak ziurtatzen du bi guneetako gasen oreka mekanikoa: ezkerrekoa zabaltzen da eskuinekoa konprimituz eta bien presioa berdina izanik.   
<br>
 5. Sistema konposatuak ez dauka ukipen mekanikorik ingurunearekin, beraz, itxitura-baldintza da ez dagoela kanpo-lanik.  
<br>
 6. Azken bi horien ondorioz, sistema konposatuak egiten du lanik egiten: lan osoa nulua da, baina bi atal dauzka, era diferentzialean idatzita, prozesu kuasiestatikoa baita: $\delta W^{konposatua} = \delta W^{ez} + \delta W^{es}\Rightarrow 0 = \delta W^{ez} + \delta W^{es} \Rightarrow \delta W^{ez} = - \delta W^{es}$  
<br>
 7. Eskuineko guneko prozesua adiabatiko kuasiestatiko denez, lana adiabatikoa *egiten du* sistemak: ${(W^{es})}_{adiabatikoa} = \Delta U$.  
    Ondorioz, $W^{ez} = - \Delta U^{es} $.  
<br>
 8. Eta, $\Delta U^{ez}$ ezaguna denez, orduan, honako hau dugu: $\boxed{ Q^{ez} = \Delta U^{ez} + \Delta U^{es} }$.  
<br>
</div>
Beraz, guztiz bukatzeko ariketa, lortu behar dira honako bi hauek:  
<br>
<br>
1. nolakoa den $U=U(T, V)$ edo $U=U(T, p)$ (egia esateko, honekin nahikoa da!)  
<br>
$$ \left(\frac{\partial U}{\partial p}\right)_{T} =  -\frac{\kappa_{T}}{\alpha}\left(C_{p} - C_{V}\right) + pV\kappa_{T} \Rightarrow \left(\frac{\partial U}{\partial p}\right)_{T} = T \left(\frac{\partial V}{\partial T}\right)_{p} - p\left(\frac{\partial V}{\partial p}\right)_{T}\Rightarrow \left(\frac{\partial U}{\partial p}\right)_{T} =0$$
<br>
$$ \left(\frac{\partial U}{\partial V}\right)_{T} = \frac{\left(C_{p} - C_{V}\right)}{V\alpha} - p \Rightarrow \left(\frac{\partial U}{\partial V}\right)_{T} = - T\frac{1}{V} \left(\frac{\partial V}{\partial T}\right)_{p} \left(\frac{\partial V}{\partial p}\right)_{T} \Rightarrow \left(\frac{\partial U}{\partial V}\right)_{T} = 0$$  
<br>
Ondorioz: $\boxed{dU = C_{V}dT }$  
<br> 
Gas idealaren egoera-ekuazio termikoa! Zuzenean kalkulatu daiteke $\Delta U$.  
<br>
2. $\delta Q = dU - \delta W \Rightarrow \boxed{0 = dU_{sis} - \delta W}$ *ekuazio diferentzialaren* integrala, horixe baita sistema jakin horren kasuan prozesu adiabatiko kuasiestatikoen forma.  
<br>
<br>
`Iruzkina:` 
<br>
Jakina, azken emaitza, $\boxed{ Q^{ez} = \Delta U^{ez} + \Delta U^{es} }$, espero zena da. Aplikatuz gero lehen printzipioa sistema konposatuari, honako hau da azterketa:  

$$ \left[Q = \Delta U - W\right]^{konposatua} $$  

 1. $ Q^{konposatua} = Q^{ez} $   
 
 2. $ \Delta U^{konposatua} = \Delta U^{ez} + \Delta U^{es} $  
 
 3. $ W ^{konposatua} = 0$
