### Ebazpen bat  


Enuntziatuak planteatzen duen prozesua *itzulezina* da: bat-batean kentzen da bi guneak banatzen dituen bereizte-horma. Dena dela, ez da zabaltze askea, sistema konposatua $T^{bi} = T_{0}$ tenperaturako bero-iturriarekin ukipen termikoan baitago. Prozesua ez da isotermikoa, ez baita $dT$ betetzen, baina badakigu hasierako eta bukaerako oreka-egoeretan sistemaren *egoera termikoa* $T_{0}$ tenperaturaren balioak ezaugarritzen duela.   

Enuntziatuaren lehen atalaka eskatzen ditu kalkulatzea sistemaren zenbait propietateren prozesuaren ondoriozko aldaketak. Propietateak direnez, berdin zaie zein prozesutan kalkulatu, ez baitute ibilbidearekiko mendekotasunik.    

Beste alde batetik, prozesuan parte hartu dituzten aldagaitzat $V$ eta $T$ har daitezke: bereizte-horma kendutakoan $V$ *trukatzen* baitu sistemak, eta $T$, esan bezala, $T_{i}^{sis}=T_{f}^{sis}=T^{bi}=T_{0}$ dela ezaguna baita.    

Orduan, idatz ditzadan potentzial termodinamiko ($PotTer$) denak aukeratutako egokiak diren aldagai termodinamikoen funtzioan eta horien adierazpen diferentziala, kontuan hartuzl $dT$ ez duela ekarpenik egingo, $T$ konstanteko prozezuan kalkulatuko baititut aldaketa denak:  


$$\left. \begin{array}{l}
    PotTer = PotTer (V, T) \Rightarrow d\,PotTer = \left(\frac{\partial\, PotTer}{\partial V}\right)_{T}\,dV + \left(\frac{\partial\, PotTer}{\partial T}\right)_{V}\,dT \\ 
    dT = 0
    \end{array} \right\} \hspace{0.5cm} \Rightarrow \hspace{0.5cm}
    \left. \begin{array}{l}
    \left[d\,PotTer = \left(\frac{\partial\, PotTer}{\partial V}\right)_{T}\,dV\right]_{T}\\
    \end{array} \right. $$   
    
Edozein kasutan, erabili behar izango ditugu $\frac{pv}{RT} = 1 - \frac{a}{vT}\Rightarrow \boxed{ p = \frac{RT}{v} -a \frac{R}{v^{2}} }$ egoera-ekuazioaren honako bi deribatu hauek:  

1. $\left(\frac{\partial p}{\partial V}\right)_{T} = \frac{R}{v}$   
<br>  
2. $\left(\frac{\partial p}{\partial T}\right)_{V} = -\frac{RT}{v} + 2a\frac{R}{v^{3}}$   
<br>  



1. **barne-energiaren aldaketa**: 
<br>
<br>
$$\left. \begin{array}{l}
\left. \begin{array}{l}
    U = U (V, T) \Rightarrow d\,U = \left(\frac{\partial\, U}{\partial V}\right)_{T}\,dV + \left(\frac{\partial\, U}{\partial T}\right)_{V}\,dT \\ 
    \\
    dT = 0\\
        \left. \begin{array}{l}
    \\
    \\
    U = U (S, V) \Rightarrow dU = T\, dS - p \, dV\\
    \end{array} \right.
    \end{array} 
    \right\} \hspace{0.5cm} \Rightarrow \hspace{0.5cm}
    \left. \begin{array}{l}
    \left[d\,U = \left(\frac{\partial\, U}{\partial V}\right)_{T}\,dV\right]_{T}\\
    \hspace{1.25cm}\Downarrow\\
    \left. \begin{array}{l}
    \left(\frac{\partial\, U}{\partial V}\right)_{T} = 
    T\,\left(\frac{\partial S}{\partial V}\right)_{T} - 
    p\,\left(\frac{\partial V}{\partial V}\right)_{T} \\
    \\
    T\,\left(\frac{\partial S}{\partial V}\right)_{T} = T\,\left(\frac{\partial p}{\partial T}\right)_{V}\\ 
    p\,\left(\frac{\partial V}{\partial V}\right)_{T} = p\\
    \end{array} \right\}
    \hspace{0.5cm}
    \Rightarrow
    \hspace{0.5cm}
    \left(\frac{\partial\, U}{\partial V}\right)_{T} = T\,\left(\frac{\partial p}{\partial T}\right)_{V} - p
    \end{array} \right. \\[0.5cm]
    \end{array} \right.
$$  
$$ \left(\frac{\partial\, U}{\partial V}\right)_{T} = 
    T\,\frac{R}{v} - \left[\frac{RT}{v} -a \frac{R}{v^{2}}\right]
    \hspace{0.5cm}
    \Rightarrow
    \hspace{0.5cm}
    \left(\frac{\partial\, U}{\partial V}\right)_{T} = +
    a \frac{R}{v^{2}}    
    \hspace{0.5cm}
    \Rightarrow
    \hspace{0.5cm}
    \Delta U = -aR\left(\frac{1}{v_{f}} - \frac{1}{v_{i}}\right)
    \hspace{0.5cm}
    \Rightarrow
    \hspace{0.5cm}
    \boxed{ \Delta U = \frac{aR}{2v} }
$$
<br>   
<br>   
1. **entalpiaren aldaketa**: 
<br>
<br>
$$\left. \begin{array}{l}
\left. \begin{array}{l}
    H = H (V, T) \Rightarrow d\,H = \left(\frac{\partial\, H}{\partial V}\right)_{T}\,dV + \left(\frac{\partial\, H}{\partial T}\right)_{V}\,dT \\ 
    \\
    dT = 0\\
        \left. \begin{array}{l}
    \\
    \\
    H = H (S, V) \Rightarrow dH = T\, dS + V \, dp\\
    \end{array} \right.
    \end{array} 
    \right\} \hspace{0.5cm} \Rightarrow \hspace{0.5cm}
    \left. \begin{array}{l}
    \left[d\,H = \left(\frac{\partial\, H}{\partial V}\right)_{T}\,dV\right]_{T}\\
    \hspace{1.25cm}\Downarrow\\
    \left. \begin{array}{l}
    \left(\frac{\partial\, H}{\partial V}\right)_{T} = 
    T\,\left(\frac{\partial S}{\partial V}\right)_{T} + 
    V\,\left(\frac{\partial p}{\partial V}\right)_{T} \\
    \\
    T\,\left(\frac{\partial S}{\partial V}\right)_{T} = T\,\left(\frac{\partial p}{\partial T}\right)_{V} \\ 
    V\,\left(\frac{\partial p}{\partial V}\right)_{T} = V\,\left(\frac{\partial p}{\partial V}\right)_{T}\\
    \end{array} \right\}
    \hspace{0.5cm}
    \Rightarrow
    \hspace{0.5cm}
    \left(\frac{\partial\, H}{\partial V}\right)_{T} = T\,\left(\frac{\partial p}{\partial T}\right)_{V} + V\,\left(\frac{\partial p}{\partial V}\right)_{T}
    \end{array} \right. \\[0.5cm]
    \end{array} \right.
$$  
$$ \left(\frac{\partial\, H}{\partial V}\right)_{T} = 
    T\,\frac{R}{v} + v\left[-\frac{RT}{v^{2}} + 2 a \frac{R}{v^{3}}\right]
    \hspace{0.5cm}
    \Rightarrow
    \hspace{0.5cm}
    \left(\frac{\partial\, H}{\partial V}\right)_{T} = -
    2 a \frac{R}{v^{3}}    
    \hspace{0.5cm}
    \Rightarrow
    \hspace{0.5cm}
    \Delta H = -2aR\left(\frac{1}{v_{f}} - \frac{1}{v_{i}}\right)
    \hspace{0.5cm}
    \Rightarrow
    \hspace{0.5cm}
    \boxed{ \Delta H = \frac{aR}{v} }
$$

1. **Helmholtz-en energia askearen aldaketa**: 
<br>
<br>
$$\left. \begin{array}{l}
\left. \begin{array}{l}
    F = F (V, T) \Rightarrow d\,F = \left(\frac{\partial\, F}{\partial V}\right)_{T}\,dV + \left(\frac{\partial\, F}{\partial T}\right)_{V}\,dT \\ 
    \\
    dT = 0\\
        \left. \begin{array}{l}
    \\
    \\
    F = F (S, V) \Rightarrow dF = -S\, dT - p \, dV\\
    \end{array} \right.
    \end{array} 
    \right\} \hspace{0.5cm} \Rightarrow \hspace{0.5cm}
    \left. \begin{array}{l}
    \left[d\,F = \left(\frac{\partial\, F}{\partial V}\right)_{T}\,dV\right]_{T}\\
    \hspace{1.25cm}\Downarrow\\
    \left. \begin{array}{l}
    \left(\frac{\partial\, F}{\partial V}\right)_{T} = -
    S\,\left(\frac{\partial T}{\partial V}\right)_{T} - 
    p\,\left(\frac{\partial V}{\partial V}\right)_{T} \\
    \\
    S\,\left(\frac{\partial T}{\partial V}\right)_{T} = 0\\ 
    p\,\left(\frac{\partial V}{\partial V}\right)_{T} = p\\
    \end{array} \right\}
    \hspace{0.5cm}
    \Rightarrow
    \hspace{0.5cm}
    \left(\frac{\partial\, F}{\partial V}\right)_{T} = - p
    \end{array} \right. \\[0.5cm]
    \end{array} \right.
$$  
$$ \left(\frac{\partial\, F}{\partial V}\right)_{T} = -
    \left[\frac{RT}{v} - a \frac{R}{v^{2}}\right]
    \hspace{0.5cm}
    \Rightarrow
    \hspace{0.5cm}
    \Delta F = -RT\ln\left(\frac{v_{f}}{v_{i}}\right)- aR\left(\frac{1}{v_{f}} - \frac{1}{v_{i}}\right)
    \hspace{0.5cm}
    \Rightarrow
    \hspace{0.5cm}
    \boxed{ \Delta F = -RT\ln(2) + \frac{aR}{2v} }
$$

1. **Gibbs-en energia askearen aldaketa**: 
<br>
<br>
$$\left. \begin{array}{l}
\left. \begin{array}{l}
    G = G (V, T) \Rightarrow d\,G = \left(\frac{\partial\, G}{\partial V}\right)_{T}\,dV + \left(\frac{\partial\, G}{\partial T}\right)_{V}\,dT \\ 
    \\
    dT = 0\\
        \left. \begin{array}{l}
    \\
    \\
    G = G (T, p) \Rightarrow dG = -S\, dT + V \, dp\\
    \end{array} \right.
    \end{array} 
    \right\} \hspace{0.5cm} \Rightarrow \hspace{0.5cm}
    \left. \begin{array}{l}
    \left[d\,G = \left(\frac{\partial\, G}{\partial V}\right)_{T}\,dV\right]_{T}\\
    \hspace{1.25cm}\Downarrow\\
    \left. \begin{array}{l}
    \left(\frac{\partial\, G}{\partial V}\right)_{T} = -
    S\,\left(\frac{\partial T}{\partial V}\right)_{T} + 
    V\,\left(\frac{\partial p}{\partial V}\right)_{T} \\
    \\
    S\,\left(\frac{\partial T}{\partial V}\right)_{T} = 0\\ 
    V\,\left(\frac{\partial p}{\partial V}\right)_{T} = V\,\left(\frac{\partial p}{\partial V}\right)_{T} \\
    \end{array} \right\}
    \hspace{0.5cm}
    \Rightarrow
    \hspace{0.5cm}
    \left(\frac{\partial\, G}{\partial V}\right)_{T} = V\,\left(\frac{\partial p}{\partial V}\right)_{T}
    \end{array} \right. \\[0.5cm]
    \end{array} \right.
$$  
$$ \left(\frac{\partial\, G}{\partial V}\right)_{T} = 
    V\left[-\frac{RT}{v^{2}} + 2 a \frac{R}{v^{3}}\right]
    \hspace{0.5cm}
    \Rightarrow
    \hspace{0.5cm}
    \Delta G = -RT\ln\left(\frac{v_{f}}{v_{i}}\right)- 2aR\left(\frac{1}{v_{f}} - \frac{1}{v_{i}}\right)
    \hspace{0.5cm}
    \Rightarrow
    \hspace{0.5cm}
    \boxed{ \Delta G = -RT\ln(2) +\frac{aR}{v} }
$$
